## ğŸ§ª Pratik Ã–rnekler: Temel KarÅŸÄ±laÅŸtÄ±rmalar

````python
# ğŸ”¢ DeÄŸiÅŸkenlerimizi tanÄ±mlayalÄ±m
# YBS Senaryosu: Bir e-ticaret sitesindeki iki Ã¼rÃ¼nÃ¼n fiyatlarÄ±

urun_a_fiyat = 150  # ÃœrÃ¼n A'nÄ±n fiyatÄ± (TL)
urun_b_fiyat = 200  # ÃœrÃ¼n B'nin fiyatÄ± (TL)

print("=" * 50)
print("ğŸ›ï¸  E-TÄ°CARET FÄ°YAT KARÅILAÅTIRMA SÄ°STEMÄ°")
print("=" * 50)
print(f"ğŸ“¦ ÃœrÃ¼n A FiyatÄ±: {urun_a_fiyat} TL")
print(f"ğŸ“¦ ÃœrÃ¼n B FiyatÄ±: {urun_b_fiyat} TL")
print("=" * 50)
````

### 1ï¸âƒ£ BÃ¼yÃ¼ktÃ¼r OperatÃ¶rÃ¼: `>`

````python
# > operatÃ¶rÃ¼: Soldaki deÄŸer saÄŸdakinden bÃ¼yÃ¼k mÃ¼?

print("ğŸ” KarÅŸÄ±laÅŸtÄ±rma: urun_b_fiyat > urun_a_fiyat")
print(f"ğŸ” Yani: {urun_b_fiyat} > {urun_a_fiyat}")
sonuc = urun_b_fiyat > urun_a_fiyat
print(f"âœ… SonuÃ§: {sonuc}")
print(f"ğŸ“Š Tip: {type(sonuc)}")  # Boolean tipinde
````

## ğŸ¯ 1. MÃ¼ÅŸteri Harcama Segmentasyonu

<div style="background: linear-gradient(135deg, #fa709a 0%, #fee140 100%); padding: 15px; border-radius: 10px; color: white; margin: 15px 0;">
    <b>ğŸ’° Ä°ÅŸ Problemi</b><br>
    MÃ¼ÅŸterileri toplam harcama tutarlarÄ±na gÃ¶re segmentlere ayÄ±rarak, her segment iÃ§in Ã¶zel pazarlama stratejileri geliÅŸtirmek
</div>

````python
# ğŸ’° MÃ¼ÅŸteri BazÄ±nda Toplam Harcama Hesaplama ve Segmentasyon

print("=" * 90)
print("ğŸ’° MÃœÅTERÄ° HARCAMA SEGMENTASYONU SÄ°STEMÄ°")
print("=" * 90)
print()

# Eksik CustomerID'leri temizle
df_clean = df[df['CustomerID'].notna()].copy()

print(f"ğŸ“Š TemizlenmiÅŸ Veri: {len(df_clean):,} satÄ±r (CustomerID olan kayÄ±tlar)")
print()

# Her mÃ¼ÅŸteri iÃ§in toplam harcama hesapla
musteri_harcama = df_clean.groupby('CustomerID')['TotalPrice'].sum().reset_index()
musteri_harcama.columns = ['CustomerID', 'ToplamHarcama']

# Ä°statistikleri gÃ¶ster
print("ğŸ“ˆ HARCAMA Ä°STATÄ°STÄ°KLERÄ°:")
print(f"   ğŸ’ Ortalama Harcama: Â£{musteri_harcama['ToplamHarcama'].mean():,.2f}")
print(f"   ğŸ“Š Medyan Harcama: Â£{musteri_harcama['ToplamHarcama'].median():,.2f}")
print(f"   ğŸ” En YÃ¼ksek Harcama: Â£{musteri_harcama['ToplamHarcama'].max():,.2f}")
print(f"   ğŸ”» En DÃ¼ÅŸÃ¼k Harcama: Â£{musteri_harcama['ToplamHarcama'].min():,.2f}")
print()
print("â”€" * 90)
print()

# IF-ELIF-ELSE ile mÃ¼ÅŸteri segmentasyonu
def musteri_segmenti_belirle(harcama):
    """
    MÃ¼ÅŸteri harcamasÄ±na gÃ¶re segment belirler
    """
    if harcama >= 10000:
        return "ğŸ’ VIP"
    elif harcama >= 5000:
        return "ğŸ¥‡ GOLD"
    elif harcama >= 2000:
        return "ğŸ¥ˆ SILVER"
    elif harcama >= 500:
        return "ğŸ¥‰ BRONZ"
    else:
        return "ğŸ”µ STANDART"

# Segment sÃ¼tununu oluÅŸtur
musteri_harcama['Segment'] = musteri_harcama['ToplamHarcama'].apply(musteri_segmenti_belirle)

print("ğŸ¯ MÃœÅTERÄ° SEGMENTLERÄ° OLUÅTURULDU!")
print()

# Segment daÄŸÄ±lÄ±mÄ±
segment_dagilim = musteri_harcama['Segment'].value_counts()

print("ğŸ“Š SEGMENT DAÄILIMI:")
print()
for segment, sayi in segment_dagilim.items():
    yuzde = (sayi / len(musteri_harcama)) * 100
    print(f"{segment:15s}: {sayi:5,} mÃ¼ÅŸteri ({yuzde:5.2f}%)")

print()
print("â”€" * 90)
print()

# Segment bazÄ±nda ortalama harcama
segment_ortalamalari = musteri_harcama.groupby('Segment')['ToplamHarcama'].agg(['mean', 'sum', 'count'])
segment_ortalamalari.columns = ['Ortalama', 'Toplam', 'MusteriSayisi']
segment_ortalamalari = segment_ortalamalari.sort_values('Ortalama', ascending=False)

print("ğŸ’° SEGMENT BAZINDA HARCAMA ANALÄ°ZÄ°:")
print()
for segment in segment_ortalamalari.index:
    ort = segment_ortalamalari.loc[segment, 'Ortalama']
    top = segment_ortalamalari.loc[segment, 'Toplam']
    say = int(segment_ortalamalari.loc[segment, 'MusteriSayisi'])
    
    print(f"{segment}:")
    print(f"   Ortalama Harcama: Â£{ort:,.2f}")
    print(f"   Toplam Gelir: Â£{top:,.2f}")
    print(f"   MÃ¼ÅŸteri SayÄ±sÄ±: {say:,}")
    print()

print("=" * 90)
print("ğŸ’¡ Bu segmentasyon CRM sistemlerinde mÃ¼ÅŸteri deÄŸerini Ã¶lÃ§mek iÃ§in kullanÄ±lÄ±r!")
print("=" * 90)
````

## ğŸ›ï¸ 2. ÃœrÃ¼n Fiyat Kategorilendirmesi

<div style="background: linear-gradient(135deg, #11998e 0%, #38ef7d 100%); padding: 15px; border-radius: 10px; color: white; margin: 15px 0;">
    <b>ğŸ·ï¸ Ä°ÅŸ Problemi</b><br>
    ÃœrÃ¼nleri fiyat aralÄ±klarÄ±na gÃ¶re kategorilere ayÄ±rarak, fiyat stratejileri ve stok yÃ¶netimi iÃ§in karar destek saÄŸlamak

````python
# ğŸ·ï¸ ÃœrÃ¼n Fiyat Kategorilendirmesi

print("=" * 90)
print("ğŸ·ï¸ ÃœRÃœN FÄ°YAT KATEGORÄ°LENDÄ°RME SÄ°STEMÄ°")
print("=" * 90)
print()

# Pozitif fiyatlÄ± Ã¼rÃ¼nleri al
df_positive = df_clean[df_clean['UnitPrice'] > 0].copy()

print(f"ğŸ“Š Pozitif FiyatlÄ± ÃœrÃ¼nler: {len(df_positive):,} kayÄ±t")
print()

# Fiyat istatistikleri
print("ğŸ“ˆ FÄ°YAT Ä°STATÄ°STÄ°KLERÄ°:")
print(f"   ğŸ’° Ortalama Fiyat: Â£{df_positive['UnitPrice'].mean():.2f}")
print(f"   ğŸ“Š Medyan Fiyat: Â£{df_positive['UnitPrice'].median():.2f}")
print(f"   ğŸ” En PahalÄ±: Â£{df_positive['UnitPrice'].max():,.2f}")
print(f"   ğŸ”» En Ucuz: Â£{df_positive['UnitPrice'].min():.2f}")
print()
print("â”€" * 90)
print()

# IF-ELIF-ELSE ile fiyat kategorilendirmesi
def fiyat_kategorisi_belirle(fiyat):
    """
    ÃœrÃ¼n fiyatÄ±na gÃ¶re kategori belirler
    """
    if fiyat >= 50:
        return "ğŸ’ PREMIUM"
    elif fiyat >= 20:
        return "ğŸ¥‡ YÃœKSEK"
    elif fiyat >= 5:
        return "ğŸ¥ˆ ORTA"
    elif fiyat >= 1:
        return "ğŸ¥‰ DÃœÅÃœK"
    else:
        return "ğŸ”µ Ã‡OK DÃœÅÃœK"

# Kategori sÃ¼tununu oluÅŸtur
df_positive['FiyatKategorisi'] = df_positive['UnitPrice'].apply(fiyat_kategorisi_belirle)

print("ğŸ¯ FÄ°YAT KATEGORÄ°LERÄ° OLUÅTURULDU!")
print()

# Kategori daÄŸÄ±lÄ±mÄ±
kategori_dagilim = df_positive['FiyatKategorisi'].value_counts()

print("ğŸ“Š FÄ°YAT KATEGORÄ°SÄ° DAÄILIMI:")
print()
for kategori, sayi in kategori_dagilim.items():
    yuzde = (sayi / len(df_positive)) * 100
    print(f"{kategori:15s}: {sayi:7,} Ã¼rÃ¼n ({yuzde:5.2f}%)")

print()
print("â”€" * 90)
print()

# Kategori bazÄ±nda satÄ±ÅŸ analizi
kategori_analiz = df_positive.groupby('FiyatKategorisi').agg({
    'Quantity': 'sum',
    'TotalPrice': 'sum',
    'InvoiceNo': 'count'
}).round(2)

kategori_analiz.columns = ['ToplamMiktar', 'ToplamGelir', 'SatisSayisi']
kategori_analiz = kategori_analiz.sort_values('ToplamGelir', ascending=False)

print("ğŸ’° KATEGORÄ° BAZINDA SATIÅ ANALÄ°ZÄ°:")
print()
for kategori in kategori_analiz.index:
    miktar = int(kategori_analiz.loc[kategori, 'ToplamMiktar'])
    gelir = kategori_analiz.loc[kategori, 'ToplamGelir']
    satis = int(kategori_analiz.loc[kategori, 'SatisSayisi'])
    
    print(f"{kategori}:")
    print(f"   Toplam SatÄ±ÅŸ MiktarÄ±: {miktar:,} adet")
    print(f"   Toplam Gelir: Â£{gelir:,.2f}")
    print(f"   SatÄ±ÅŸ SayÄ±sÄ±: {satis:,}")
    
    # IF ile performans deÄŸerlendirmesi
    if gelir > 1000000:
        performans = "ğŸŒŸ MÃ¼kemmel - YÃ¼ksek gelir kaynaÄŸÄ±"
    elif gelir > 500000:
        performans = "âœ… Ä°yi - Ã–nemli gelir katkÄ±sÄ±"
    else:
        performans = "âš ï¸  Normal - Potansiyel artÄ±ÅŸ alanÄ±"
    
    print(f"   Performans: {performans}")
    print()

print("=" * 90)
print("ğŸ’¡ Bu kategorilendirme fiyatlandÄ±rma stratejisi iÃ§in kullanÄ±lÄ±r!")
print("=" * 90)
````

### ğŸ¯ 9.1. RFM Analizi ile MÃ¼ÅŸteri DeÄŸeri Segmentasyonu

**RFM** (Recency, Frequency, Monetary), mÃ¼ÅŸteri deÄŸerini anlamak iÃ§in kullanÄ±lan en gÃ¼Ã§lÃ¼ analitik yÃ¶ntemlerden biridir.

<div style="background: linear-gradient(135deg, #f093fb 0%, #f5576c 100%); padding: 20px; border-radius: 10px; color: white; margin: 15px 0;">
<h4 style="margin-top: 0;">ğŸ“Œ RFM Nedir?</h4>

- **R (Recency)**: MÃ¼ÅŸteri son ne zaman alÄ±ÅŸveriÅŸ yaptÄ±?
- **F (Frequency)**: MÃ¼ÅŸteri ne sÄ±klÄ±kla alÄ±ÅŸveriÅŸ yapÄ±yor?
- **M (Monetary)**: MÃ¼ÅŸteri ne kadar harcama yapÄ±yor?

**YBS'de KullanÄ±m:** CRM sistemlerinde mÃ¼ÅŸteri sadakat programlarÄ±nÄ±n tasarÄ±mÄ± iÃ§in kritik!
</div>

````python
# ğŸ“Š RFM Analizi - IF KontrolÃ¼ ile Segmentasyon
# ================================================

# Analiz iÃ§in referans tarih belirleme (veri setindeki son tarih)
referans_tarih = df_positive['InvoiceDate'].max() + pd.Timedelta(days=1)
print(f"ğŸ—“ï¸ Referans Tarih: {referans_tarih}\n")

# RFM deÄŸerlerini hesaplama
rfm = df_positive.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (referans_tarih - x.max()).days,  # Recency
    'InvoiceNo': 'count',                                       # Frequency
    'TotalPrice': 'sum'                                         # Monetary
}).reset_index()

# Kolon isimlerini dÃ¼zenleme
rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']

print("ğŸ“ˆ RFM DeÄŸerleri Ä°lk 10 MÃ¼ÅŸteri:")
print(rfm.head(10))
print(f"\nğŸ“Š Toplam MÃ¼ÅŸteri SayÄ±sÄ±: {len(rfm)}")

# RFM istatistikleri
print("\n" + "="*70)
print("ğŸ“Š RFM Ä°STATÄ°STÄ°KLERÄ°")
print("="*70)
print(f"â€¢ Recency OrtalamasÄ±: {rfm['Recency'].mean():.1f} gÃ¼n")
print(f"â€¢ Frequency OrtalamasÄ±: {rfm['Frequency'].mean():.1f} sipariÅŸ")
print(f"â€¢ Monetary OrtalamasÄ±: Â£{rfm['Monetary'].mean():.2f}")
print("="*70)
````

````python
# ğŸ¯ RFM SkorlamasÄ± - IF-ELIF-ELSE ile Ã‡ok KatmanlÄ± DeÄŸerlendirme
# ==================================================================

def rfm_skoru_hesapla(recency, frequency, monetary):
    """
    RFM metriklerine gÃ¶re mÃ¼ÅŸteriye 1-5 arasÄ± skor verir.
    
    MantÄ±k:
    - Recency: DÃ¼ÅŸÃ¼k = Ä°YÄ° (yakÄ±n zamanda alÄ±ÅŸveriÅŸ yapmÄ±ÅŸ)
    - Frequency: YÃ¼ksek = Ä°YÄ° (sÄ±k alÄ±ÅŸveriÅŸ yapÄ±yor)
    - Monetary: YÃ¼ksek = Ä°YÄ° (Ã§ok harcama yapÄ±yor)
    """
    
    # Recency SkorlamasÄ± (DÃ¼ÅŸÃ¼k recency = YÃ¼ksek skor)
    if recency <= 30:
        r_score = 5    # MÃ¼kemmel! Son 30 gÃ¼nde gelmiÅŸ
    elif recency <= 60:
        r_score = 4    # Ã‡ok iyi! Son 2 ay iÃ§inde gelmiÅŸ
    elif recency <= 90:
        r_score = 3    # Ä°yi! Son 3 ay iÃ§inde gelmiÅŸ
    elif recency <= 180:
        r_score = 2    # Orta! Son 6 ay iÃ§inde gelmiÅŸ
    else:
        r_score = 1    # KÃ¶tÃ¼! 6 aydan uzun sÃ¼re gelmemiÅŸ
    
    # Frequency SkorlamasÄ± (YÃ¼ksek frequency = YÃ¼ksek skor)
    if frequency >= 50:
        f_score = 5    # SÃ¼per sadÄ±k mÃ¼ÅŸteri!
    elif frequency >= 20:
        f_score = 4    # Ã‡ok sadÄ±k mÃ¼ÅŸteri
    elif frequency >= 10:
        f_score = 3    # SadÄ±k mÃ¼ÅŸteri
    elif frequency >= 5:
        f_score = 2    # Orta seviye mÃ¼ÅŸteri
    else:
        f_score = 1    # Yeni veya dÃ¼ÅŸÃ¼k frekanslÄ± mÃ¼ÅŸteri
    
    # Monetary SkorlamasÄ± (YÃ¼ksek monetary = YÃ¼ksek skor)
    if monetary >= 5000:
        m_score = 5    # Premium harcayÄ±cÄ±!
    elif monetary >= 2000:
        m_score = 4    # YÃ¼ksek harcayÄ±cÄ±
    elif monetary >= 1000:
        m_score = 3    # Orta harcayÄ±cÄ±
    elif monetary >= 500:
        m_score = 2    # DÃ¼ÅŸÃ¼k harcayÄ±cÄ±
    else:
        m_score = 1    # Minimal harcayÄ±cÄ±
    
    return r_score, f_score, m_score


# RFM skorlarÄ±nÄ± hesaplama
rfm[['R_Score', 'F_Score', 'M_Score']] = rfm.apply(
    lambda row: rfm_skoru_hesapla(row['Recency'], row['Frequency'], row['Monetary']), 
    axis=1, 
    result_type='expand'
)

# RFM Toplam Skoru (1-15 arasÄ±)
rfm['RFM_Score'] = rfm['R_Score'] + rfm['F_Score'] + rfm['M_Score']

print("ğŸ† RFM SKORLARI (Ä°lk 15 MÃ¼ÅŸteri)")
print("="*90)
print(rfm[['CustomerID', 'Recency', 'Frequency', 'Monetary', 'R_Score', 'F_Score', 'M_Score', 'RFM_Score']].head(15))
print("="*90)
````

````python
# ğŸ¯ RFM Segmentleri - IF-ELIF-ELSE ile Stratejik Kategorilendirme
# ===================================================================

def rfm_segment_belirle(rfm_score, r_score, f_score, m_score):
    """
    RFM skorlarÄ±na gÃ¶re stratejik mÃ¼ÅŸteri segmentleri oluÅŸturur.
    YBS'de bu segmentler, farklÄ± pazarlama stratejileri iÃ§in kullanÄ±lÄ±r.
    """
    
    # Champions (Åampiyonlar) - En deÄŸerli mÃ¼ÅŸteriler
    if rfm_score >= 13 and r_score >= 4 and f_score >= 4:
        return "ğŸ† Champions"
    
    # Loyal Customers (SadÄ±k MÃ¼ÅŸteriler) - YÃ¼ksek frequency
    elif f_score >= 4 and m_score >= 3:
        return "ğŸ’ Loyal Customers"
    
    # Potential Loyalists (Potansiyel SadÄ±klar) - Son zamanda gelmiÅŸ, artÄ±rma potansiyeli var
    elif r_score >= 4 and f_score >= 2 and m_score >= 2:
        return "ğŸŒŸ Potential Loyalists"
    
    # New Customers (Yeni MÃ¼ÅŸteriler) - YakÄ±n zamanda gelmiÅŸ ama az harcamÄ±ÅŸ
    elif r_score >= 4 and f_score <= 2:
        return "ğŸ†• New Customers"
    
    # At Risk (Risk AltÄ±nda) - Ã–nceden iyiydi ama son zamanda gelmedi
    elif r_score <= 2 and f_score >= 3:
        return "âš ï¸ At Risk"
    
    # Can't Lose Them (Kaybetme Riski) - YÃ¼ksek harcayÄ±cÄ± ama gelmemiÅŸ
    elif r_score <= 2 and m_score >= 4:
        return "ğŸš¨ Can't Lose Them"
    
    # Hibernating (Uykuda) - Uzun sÃ¼redir gelmemiÅŸ
    elif r_score <= 2 and f_score <= 2:
        return "ğŸ˜´ Hibernating"
    
    # Promising (Umut Vaat Eden) - Orta seviye ama potansiyel var
    elif rfm_score >= 7 and rfm_score < 10:
        return "ğŸŒ± Promising"
    
    # Others (DiÄŸerleri) - Standart mÃ¼ÅŸteriler
    else:
        return "ğŸ“Š Others"


# Segmentleri belirleme
rfm['Segment'] = rfm.apply(
    lambda row: rfm_segment_belirle(row['RFM_Score'], row['R_Score'], row['F_Score'], row['M_Score']),
    axis=1
)

print("\n" + "="*90)
print("ğŸ¯ RFM SEGMENT ANALÄ°ZÄ° - DETAYLI RAPOR")
print("="*90)

# Segment bazÄ±nda analiz
segment_analiz = rfm.groupby('Segment').agg({
    'CustomerID': 'count',
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': 'mean',
    'RFM_Score': 'mean'
}).round(2)

segment_analiz.columns = ['MÃ¼ÅŸteri SayÄ±sÄ±', 'Ort. Recency', 'Ort. Frequency', 'Ort. Monetary', 'Ort. RFM Skor']
segment_analiz = segment_analiz.sort_values('MÃ¼ÅŸteri SayÄ±sÄ±', ascending=False)

print(segment_analiz)
print("="*90)

# Segment daÄŸÄ±lÄ±mÄ±
print("\nğŸ“Š SEGMENT DAÄILIMI:")
print(rfm['Segment'].value_counts())
````

````python
# ğŸ¯ Segment BazlÄ± Pazarlama Stratejisi - IF-ELIF-ELSE Karar Sistemi
# ======================================================================

def pazarlama_stratejisi_belirle(segment):
    """
    Her RFM segmenti iÃ§in Ã¶zel pazarlama stratejisi ve eylem planÄ± belirler.
    YBS'de bu, CRM otomasyonu iÃ§in kullanÄ±lan kritik bir iÅŸ kuralÄ±dÄ±r.
    """
    
    if segment == "ğŸ† Champions":
        strateji = "VIP ProgramÄ± & Ã–zel Teklifler"
        eylem = "- Erken eriÅŸim kampanyalarÄ±\n  - Referans bonuslarÄ±\n  - KiÅŸiselleÅŸtirilmiÅŸ hizmet"
        butce = "YÃ¼ksek"
        oncelik = "â­â­â­â­â­"
    
    elif segment == "ğŸ’ Loyal Customers":
        strateji = "Sadakat ProgramÄ± GÃ¼Ã§lendirme"
        eylem = "- Puan sistemi\n  - DoÄŸum gÃ¼nÃ¼ indirimleri\n  - Cross-sell kampanyalarÄ±"
        butce = "YÃ¼ksek"
        oncelik = "â­â­â­â­"
    
    elif segment == "ğŸŒŸ Potential Loyalists":
        strateji = "Sadakat ArtÄ±rma KampanyalarÄ±"
        eylem = "- Ãœcretsiz kargo\n  - Combo teklifler\n  - Membership programlarÄ±"
        butce = "Orta"
        oncelik = "â­â­â­â­"
    
    elif segment == "ğŸ†• New Customers":
        strateji = "HoÅŸ Geldin KampanyalarÄ±"
        eylem = "- Ä°lk alÄ±ÅŸveriÅŸ indirimi\n  - ÃœrÃ¼n Ã¶nerileri\n  - Email nurturing"
        butce = "Orta"
        oncelik = "â­â­â­"
    
    elif segment == "âš ï¸ At Risk":
        strateji = "Geri Kazanma KampanyalarÄ±"
        eylem = "- Ã–zel indirim kodlarÄ±\n  - 'Ã–zlendik' mesajlarÄ±\n  - Anket ve feedback"
        butce = "YÃ¼ksek"
        oncelik = "â­â­â­â­"
    
    elif segment == "ğŸš¨ Can't Lose Them":
        strateji = "ACÄ°L Geri Kazanma"
        eylem = "- KiÅŸisel arama\n  - Ã–zel indirimler\n  - Problem Ã§Ã¶zme odaklÄ± yaklaÅŸÄ±m"
        butce = "Ã‡ok YÃ¼ksek"
        oncelik = "â­â­â­â­â­"
    
    elif segment == "ğŸ˜´ Hibernating":
        strateji = "Reaktivasyon KampanyalarÄ±"
        eylem = "- Agresif indirimler\n  - Yeni Ã¼rÃ¼n tanÄ±tÄ±mlarÄ±\n  - Win-back emails"
        butce = "DÃ¼ÅŸÃ¼k-Orta"
        oncelik = "â­â­"
    
    elif segment == "ğŸŒ± Promising":
        strateji = "BÃ¼yÃ¼me Destekleme"
        eylem = "- Bundle teklifler\n  - Upsell fÄ±rsatlarÄ±\n  - EÄŸitici iÃ§erik"
        butce = "Orta"
        oncelik = "â­â­â­"
    
    else:  # Others
        strateji = "Standart Pazarlama"
        eylem = "- Genel kampanyalar\n  - Newsletter\n  - Sezonsal teklifler"
        butce = "DÃ¼ÅŸÃ¼k"
        oncelik = "â­â­"
    
    return strateji, eylem, butce, oncelik


# Her segment iÃ§in stratejileri belirleme
print("\n" + "="*100)
print("ğŸ¯ SEGMENT BAZLI PAZARLAMAStrategy PLANI - YBS CRM SÄ°STEMÄ°")
print("="*100)

for segment in rfm['Segment'].unique():
    strateji, eylem, butce, oncelik = pazarlama_stratejisi_belirle(segment)
    musteri_sayisi = len(rfm[rfm['Segment'] == segment])
    
    print(f"\n{segment} ({musteri_sayisi} mÃ¼ÅŸteri)")
    print(f"  ğŸ“‹ Strateji: {strateji}")
    print(f"  âœ… Eylemler:\n  {eylem}")
    print(f"  ğŸ’° BÃ¼tÃ§e: {butce}")
    print(f"  ğŸ¯ Ã–ncelik: {oncelik}")
    print("-" * 100)
````

- **HaftanÄ±n GÃ¼nÃ¼**: Hangi gÃ¼nler daha aktif?

````python
# ğŸ• Zaman TabanlÄ± Ã–zellikler OluÅŸturma
# ========================================

# Tarih sÃ¼tunundan zaman Ã¶zelliklerini Ã§Ä±karma
df_positive['Saat'] = df_positive['InvoiceDate'].dt.hour
df_positive['Gun'] = df_positive['InvoiceDate'].dt.day_name()
df_positive['Ay'] = df_positive['InvoiceDate'].dt.month
df_positive['AyAdi'] = df_positive['InvoiceDate'].dt.month_name()
df_positive['Ceyrek'] = df_positive['InvoiceDate'].dt.quarter
df_positive['HaftaninGunu'] = df_positive['InvoiceDate'].dt.dayofweek  # 0=Monday, 6=Sunday

print("ğŸ• ZAMAN BAZLI Ã–ZELLÄ°KLER")
print("="*70)
print(df_positive[['InvoiceDate', 'Saat', 'Gun', 'AyAdi', 'Ceyrek', 'TotalPrice']].head(10))
print("="*70)
````

````python
# ğŸ• SipariÅŸ Saati Segmentasyonu - IF-ELIF-ELSE ile Zaman Dilimleri
# ====================================================================

def saat_dilimi_belirle(saat):
    """
    SipariÅŸ saatine gÃ¶re gÃ¼nÃ¼n hangi dÃ¶neminde iÅŸlem yapÄ±ldÄ±ÄŸÄ±nÄ± belirler.
    YBS'de bu, operasyonel planlama ve mÃ¼ÅŸteri hizmetleri iÃ§in kullanÄ±lÄ±r.
    """
    
    if saat >= 6 and saat < 12:
        return "ğŸŒ… Sabah (06:00-12:00)"
    elif saat >= 12 and saat < 18:
        return "â˜€ï¸ Ã–ÄŸle (12:00-18:00)"
    elif saat >= 18 and saat < 22:
        return "ğŸŒ† AkÅŸam (18:00-22:00)"
    else:  # 22:00-06:00
        return "ğŸŒ™ Gece (22:00-06:00)"


# Saat dilimlerini belirleme
df_positive['SaatDilimi'] = df_positive['Saat'].apply(saat_dilimi_belirle)

# Saat dilimi analizi
saat_analiz = df_positive.groupby('SaatDilimi').agg({
    'InvoiceNo': 'count',
    'TotalPrice': 'sum',
    'Quantity': 'sum'
}).round(2)

saat_analiz.columns = ['SipariÅŸ SayÄ±sÄ±', 'Toplam Gelir (Â£)', 'Toplam ÃœrÃ¼n']
saat_analiz['Ort. SipariÅŸ DeÄŸeri (Â£)'] = (saat_analiz['Toplam Gelir (Â£)'] / saat_analiz['SipariÅŸ SayÄ±sÄ±']).round(2)

print("\n" + "="*100)
print("ğŸ• SAAT DÄ°LÄ°MÄ° ANALÄ°ZÄ° - OPERASYONEL PLANLAMA Ä°Ã‡Ä°N")
print("="*100)
print(saat_analiz.sort_values('SipariÅŸ SayÄ±sÄ±', ascending=False))
print("="*100)

# En yoÄŸun saat dilimini bulma
en_yogun_dilim = saat_analiz['SipariÅŸ SayÄ±sÄ±'].idxmax()
print(f"\nğŸ† En YoÄŸun Saat Dilimi: {en_yogun_dilim}")
print(f"ğŸ“Š Toplam SipariÅŸlerin %{(saat_analiz.loc[en_yogun_dilim, 'SipariÅŸ SayÄ±sÄ±'] / saat_analiz['SipariÅŸ SayÄ±sÄ±'].sum() * 100):.1f}'si bu dilimde!")
````

````python
# ğŸ“… Hafta Ä°Ã§i vs Hafta Sonu Analizi - IF-ELSE ile Ä°ÅŸ GÃ¼nÃ¼ Segmentasyonu
# ==========================================================================

def gun_tipi_belirle(hafta_gunu):
    """
    HaftanÄ±n gÃ¼nÃ¼ne gÃ¶re iÅŸ gÃ¼nÃ¼ mÃ¼ hafta sonu mu olduÄŸunu belirler.
    
    0 = Monday, 1 = Tuesday, ..., 6 = Sunday
    """
    
    if hafta_gunu < 5:  # 0-4: Monday to Friday
        return "ğŸ’¼ Hafta Ä°Ã§i"
    else:  # 5-6: Saturday, Sunday
        return "ğŸ‰ Hafta Sonu"


# GÃ¼n tipini belirleme
df_positive['GunTipi'] = df_positive['HaftaninGunu'].apply(gun_tipi_belirle)

# GÃ¼n tipi analizi
gun_tipi_analiz = df_positive.groupby('GunTipi').agg({
    'InvoiceNo': 'count',
    'TotalPrice': ['sum', 'mean'],
    'CustomerID': 'nunique'
}).round(2)

gun_tipi_analiz.columns = ['SipariÅŸ SayÄ±sÄ±', 'Toplam Gelir (Â£)', 'Ort. SipariÅŸ (Â£)', 'Benzersiz MÃ¼ÅŸteri']

print("\n" + "="*90)
print("ğŸ“… HAFTA Ä°Ã‡Ä° vs HAFTA SONU ANALÄ°ZÄ°")
print("="*90)
print(gun_tipi_analiz)
print("="*90)

# Hafta iÃ§i vs hafta sonu karÅŸÄ±laÅŸtÄ±rmasÄ±
hafta_ici = gun_tipi_analiz.loc['ğŸ’¼ Hafta Ä°Ã§i', 'SipariÅŸ SayÄ±sÄ±']
hafta_sonu = gun_tipi_analiz.loc['ğŸ‰ Hafta Sonu', 'SipariÅŸ SayÄ±sÄ±']

if hafta_ici > hafta_sonu:
    fark = ((hafta_ici - hafta_sonu) / hafta_sonu * 100)
    print(f"\nğŸ’¼ Hafta iÃ§i sipariÅŸler, hafta sonu sipariÅŸlerinden %{fark:.1f} daha fazla!")
    print("ğŸ“Š Strateji: Hafta iÃ§i kampanyalara odaklanÄ±n.")
else:
    fark = ((hafta_sonu - hafta_ici) / hafta_ici * 100)
    print(f"\nğŸ‰ Hafta sonu sipariÅŸler, hafta iÃ§i sipariÅŸlerden %{fark:.1f} daha fazla!")
    print("ğŸ“Š Strateji: Hafta sonu Ã¶zel kampanyalar dÃ¼zenleyin.")
````

````python
# ğŸ“† Mevsimsel Analiz - IF-ELIF-ELSE ile Ã‡eyrek ve Ay BazlÄ± Segmentasyon
# ==========================================================================

def mevsim_belirle(ay):
    """
    Aya gÃ¶re mevsimi belirler (Kuzey YarÄ±mkÃ¼re standartlarÄ±).
    YBS'de mevsimsel trend analizi iÃ§in kullanÄ±lÄ±r.
    """
    
    if ay in [12, 1, 2]:
        return "â„ï¸ KÄ±ÅŸ"
    elif ay in [3, 4, 5]:
        return "ğŸŒ¸ Ä°lkbahar"
    elif ay in [6, 7, 8]:
        return "â˜€ï¸ Yaz"
    else:  # 9, 10, 11
        return "ğŸ‚ Sonbahar"


# Mevsim belirleme
df_positive['Mevsim'] = df_positive['Ay'].apply(mevsim_belirle)

# Mevsimsel analiz
mevsim_analiz = df_positive.groupby('Mevsim').agg({
    'InvoiceNo': 'count',
    'TotalPrice': 'sum',
    'CustomerID': 'nunique'
}).round(2)

mevsim_analiz.columns = ['SipariÅŸ SayÄ±sÄ±', 'Toplam Gelir (Â£)', 'Benzersiz MÃ¼ÅŸteri']
mevsim_analiz['MÃ¼ÅŸteri BaÅŸÄ± Harcama (Â£)'] = (mevsim_analiz['Toplam Gelir (Â£)'] / mevsim_analiz['Benzersiz MÃ¼ÅŸteri']).round(2)

# Mevsim sÄ±ralamasÄ± (KÄ±ÅŸ, Ä°lkbahar, Yaz, Sonbahar)
mevsim_sirasi = ["â„ï¸ KÄ±ÅŸ", "ğŸŒ¸ Ä°lkbahar", "â˜€ï¸ Yaz", "ğŸ‚ Sonbahar"]
mevsim_analiz = mevsim_analiz.reindex(mevsim_sirasi)

print("\n" + "="*100)
print("ğŸŒ MEVSÄ°MSEL TREND ANALÄ°ZÄ° - YILLIK PLANLAMA Ä°Ã‡Ä°N")
print("="*100)
print(mevsim_analiz)
print("="*100)

# En karlÄ± mevsimi bulma
en_karli_mevsim = mevsim_analiz['Toplam Gelir (Â£)'].idxmax()
en_yuksek_gelir = mevsim_analiz['Toplam Gelir (Â£)'].max()

print(f"\nğŸ† En KarlÄ± Mevsim: {en_karli_mevsim}")
print(f"ğŸ’° Toplam Gelir: Â£{en_yuksek_gelir:,.2f}")
print(f"ğŸ“Š YÄ±llÄ±k Gelirin %{(en_yuksek_gelir / mevsim_analiz['Toplam Gelir (Â£)'].sum() * 100):.1f}'si bu mevsimde gerÃ§ekleÅŸiyor!")
````

**YBS'de KullanÄ±m:** Veri kalitesi kontrolÃ¼, fraud detection, sistem izleme iÃ§in vazgeÃ§ilmez!

````python
# ğŸš¨ Ä°statistiksel Anomali Tespiti - IF-ELSE ile Outlier Detection
# ===================================================================

# TotalPrice iÃ§in istatistiksel eÅŸik deÄŸerlerini hesaplama
Q1 = df_positive['TotalPrice'].quantile(0.25)
Q3 = df_positive['TotalPrice'].quantile(0.75)
IQR = Q3 - Q1

# Outlier eÅŸikleri (IQR yÃ¶ntemi)
alt_esik = Q1 - 1.5 * IQR
ust_esik = Q3 + 1.5 * IQR

print("ğŸ“Š ANOMALÄ° TESPÄ°TÄ° Ä°Ã‡Ä°N Ä°STATÄ°STÄ°KSEL EÅÄ°KLER")
print("="*70)
print(f"â€¢ Q1 (1. Ã‡eyrek): Â£{Q1:.2f}")
print(f"â€¢ Q3 (3. Ã‡eyrek): Â£{Q3:.2f}")
print(f"â€¢ IQR (Ã‡eyrekler ArasÄ± AralÄ±k): Â£{IQR:.2f}")
print(f"â€¢ Alt EÅŸik: Â£{alt_esik:.2f}")
print(f"â€¢ Ãœst EÅŸik: Â£{ust_esik:.2f}")
print("="*70)


def anomali_tespit_et(fiyat, alt_esik, ust_esik):
    """
    FiyatÄ±n normal aralÄ±kta olup olmadÄ±ÄŸÄ±nÄ± kontrol eder.
    IF-ELSE mantÄ±ÄŸÄ± ile outlier detection.
    """
    
    if fiyat < alt_esik:
        return "âš ï¸ Anormal DÃ¼ÅŸÃ¼k"
    elif fiyat > ust_esik:
        return "ğŸš¨ Anormal YÃ¼ksek"
    else:
        return "âœ… Normal"


# Anomali tespiti
df_positive['AnomaliDurumu'] = df_positive['TotalPrice'].apply(
    lambda x: anomali_tespit_et(x, alt_esik, ust_esik)
)

# Anomali daÄŸÄ±lÄ±mÄ±
anomali_dagilim = df_positive['AnomaliDurumu'].value_counts()

print("\nğŸ” ANOMALÄ° DAÄILIMI:")
print("="*70)
for durum, sayi in anomali_dagilim.items():
    yuzde = (sayi / len(df_positive) * 100)
    print(f"{durum}: {sayi:,} sipariÅŸ (%{yuzde:.2f})")
print("="*70)
````

````python
# ğŸ” Anormal Ä°ÅŸlemlerin DetaylÄ± Ä°ncelenmesi - Multi-Criteria IF Kontrolleri
# =============================================================================

def anomali_risk_seviyesi_belirle(fiyat, miktar, ust_esik):
    """
    Ã‡oklu kriterlere gÃ¶re anomali risk seviyesini belirler.
    YBS'de bu, fraud detection sistemlerinde kullanÄ±lÄ±r.
    """
    
    # Ã‡ok yÃ¼ksek fiyat + Ã‡ok yÃ¼ksek miktar = Kritik Risk
    if fiyat > ust_esik * 2 and miktar > 100:
        return "ğŸ”´ KRÄ°TÄ°K RÄ°SK"
    
    # Ã‡ok yÃ¼ksek fiyat + Normal miktar = YÃ¼ksek Risk
    elif fiyat > ust_esik * 2:
        return "ğŸŸ  YÃœKSEK RÄ°SK"
    
    # YÃ¼ksek fiyat + YÃ¼ksek miktar = Orta Risk
    elif fiyat > ust_esik and miktar > 50:
        return "ğŸŸ¡ ORTA RÄ°SK"
    
    # Sadece yÃ¼ksek fiyat = DÃ¼ÅŸÃ¼k Risk
    elif fiyat > ust_esik:
        return "ğŸŸ¢ DÃœÅÃœK RÄ°SK"
    
    # Normal durumlar
    else:
        return "âœ… Normal"


# Risk seviyesi belirleme (sadece anormal yÃ¼ksek iÅŸlemler iÃ§in)
anormal_yuksek = df_positive[df_positive['AnomaliDurumu'] == 'ğŸš¨ Anormal YÃ¼ksek'].copy()
anormal_yuksek['RiskSeviyesi'] = anormal_yuksek.apply(
    lambda row: anomali_risk_seviyesi_belirle(row['TotalPrice'], row['Quantity'], ust_esik),
    axis=1
)

print("\n" + "="*100)
print("ğŸš¨ YÃœKSEK DEÄERLI ANOMALÄ° ANALÄ°ZÄ° - FRAUD DETECTION")
print("="*100)

if len(anormal_yuksek) > 0:
    # Risk seviyesi daÄŸÄ±lÄ±mÄ±
    risk_dagilim = anormal_yuksek['RiskSeviyesi'].value_counts()
    print("\nğŸ“Š Risk Seviyesi DaÄŸÄ±lÄ±mÄ±:")
    for risk, sayi in risk_dagilim.items():
        print(f"  {risk}: {sayi:,} iÅŸlem")
    
    # En yÃ¼ksek riskli iÅŸlemleri gÃ¶sterme
    print("\nğŸ”´ KRÄ°TÄ°K VE YÃœKSEK RÄ°SKLÄ° Ä°ÅLEMLER (Ä°lk 10):")
    print("="*100)
    kritik_islemler = anormal_yuksek[
        anormal_yuksek['RiskSeviyesi'].isin(['ğŸ”´ KRÄ°TÄ°K RÄ°SK', 'ğŸŸ  YÃœKSEK RÄ°SK'])
    ].nlargest(10, 'TotalPrice')
    
    for idx, row in kritik_islemler.iterrows():
        print(f"\n{row['RiskSeviyesi']}")
        print(f"  â€¢ Fatura No: {row['InvoiceNo']}")
        print(f"  â€¢ MÃ¼ÅŸteri ID: {row['CustomerID']}")
        print(f"  â€¢ ÃœrÃ¼n: {row['Description'][:50]}...")
        print(f"  â€¢ Miktar: {row['Quantity']}")
        print(f"  â€¢ Toplam Fiyat: Â£{row['TotalPrice']:,.2f}")
        print(f"  â€¢ Ãœlke: {row['Country']}")
        print(f"  â€¢ Tarih: {row['InvoiceDate']}")
        print("-" * 100)
    
    # Ä°ÅŸ kuralÄ± Ã¶nerisi
    print("\nğŸ’¡ YBS Ä°Å KURALI Ã–NERÄ°SÄ°:")
    print("="*100)
    print("1. ğŸ”´ KRÄ°TÄ°K RÄ°SK â†’ Manuel onay gerektirir")
    print("2. ğŸŸ  YÃœKSEK RÄ°SK â†’ Otomatik doÄŸrulama + bildirim")
    print("3. ğŸŸ¡ ORTA RÄ°SK â†’ Ä°ÅŸlem loglanÄ±r, periyodik incelenir")
    print("4. ğŸŸ¢ DÃœÅÃœK RÄ°SK â†’ Normal iÅŸlem akÄ±ÅŸÄ±")
    print("="*100)
else:
    print("\nâœ… Kritik riskli iÅŸlem bulunamadÄ±!")
````

### ğŸ“ 9.4. Ã‡oklu Kriter Karar Sistemi - Nested IF Controls

**Nested IF** (Ä°Ã§ iÃ§e IF), karmaÅŸÄ±k iÅŸ kurallarÄ±nÄ± modellemek iÃ§in kullanÄ±lÄ±r.

<div style="background: linear-gradient(135deg, #a8edea 0%, #fed6e3 100%); padding: 20px; border-radius: 10px; color: #333; margin: 15px 0;">
<h4 style="margin-top: 0;">ğŸ”— Nested IF Nedir?</h4>

**Ã‡oklu koÅŸullarÄ±n** birleÅŸtirilmesiyle oluÅŸan karmaÅŸÄ±k karar yapÄ±larÄ±:
- IF iÃ§inde IF (iÃ§ iÃ§e koÅŸullar)
- Ã–ncelik sÄ±rasÄ±na gÃ¶re deÄŸerlendirme
- HiyerarÅŸik karar aÄŸaÃ§larÄ±

**YBS'de KullanÄ±m:** Kredi onay sistemleri, fiyatlandÄ±rma motorlarÄ±, risk deÄŸerlendirme!
</div>

````python
# ğŸ¯ GeliÅŸmiÅŸ MÃ¼ÅŸteri SÄ±nÄ±flandÄ±rma - Nested IF ile Ã‡ok KatmanlÄ± Analiz
# ==========================================================================

def musteri_siniflandir_ileri_seviye(recency, frequency, monetary, country):
    """
    Ã‡oklu kriterlere gÃ¶re mÃ¼ÅŸteri sÄ±nÄ±flandÄ±rmasÄ±.
    NESTED IF yapÄ±sÄ± ile karmaÅŸÄ±k iÅŸ kurallarÄ±nÄ± uygular.
    
    Kriterler:
    1. CoÄŸrafi Ã¶ncelik (UK = ana pazar)
    2. RFM deÄŸerleri
    3. Harcama potansiyeli
    """
    
    # Birinci Katman: CoÄŸrafi Segmentasyon
    if country == 'United Kingdom':
        # UK mÃ¼ÅŸterileri iÃ§in Ã¶zel deÄŸerlendirme (ana pazar)
        
        if monetary > 5000:
            # YÃ¼ksek deÄŸerli UK mÃ¼ÅŸterisi
            if recency <= 30 and frequency >= 20:
                return "â­â­â­ UK VIP Elite"
            elif recency <= 60:
                return "â­â­ UK VIP Gold"
            else:
                return "â­ UK VIP Silver (Risk)"
        
        elif monetary > 2000:
            # Orta-yÃ¼ksek deÄŸerli UK mÃ¼ÅŸterisi
            if frequency >= 15:
                return "ğŸ’ UK Premium Loyal"
            elif recency <= 30:
                return "ğŸ’ UK Premium Active"
            else:
                return "ğŸ’ UK Premium (Taki Gerekir)"
        
        elif monetary > 500:
            # Standart UK mÃ¼ÅŸterisi
            if recency <= 30 and frequency >= 5:
                return "âœ… UK Standard Active"
            else:
                return "âš ï¸ UK Standard (GeliÅŸtirilmeli)"
        
        else:
            # DÃ¼ÅŸÃ¼k deÄŸerli UK mÃ¼ÅŸterisi
            return "ğŸŒ± UK Entry Level"
    
    else:
        # UluslararasÄ± mÃ¼ÅŸteriler iÃ§in deÄŸerlendirme
        
        if monetary > 3000:
            # YÃ¼ksek deÄŸerli uluslararasÄ± mÃ¼ÅŸteri
            if recency <= 60 and frequency >= 10:
                return "ğŸŒ International VIP"
            else:
                return "ğŸŒ International High Value (Risk)"
        
        elif monetary > 1000:
            # Orta deÄŸerli uluslararasÄ± mÃ¼ÅŸteri
            if frequency >= 5:
                return "ğŸŒ International Regular"
            else:
                return "ğŸŒ International Occasional"
        
        else:
            # DÃ¼ÅŸÃ¼k deÄŸerli uluslararasÄ± mÃ¼ÅŸteri
            return "ğŸŒ International Entry"


# RFM verisi ile birleÅŸtirme (country bilgisi eklenmeli)
# MÃ¼ÅŸteri baÅŸÄ±na Ã¼lke bilgisini alma
musteri_ulke = df_positive.groupby('CustomerID')['Country'].first()

# RFM'e Ã¼lke ekle
rfm_with_country = rfm.merge(
    musteri_ulke.reset_index(),
    on='CustomerID',
    how='left'
)

# GeliÅŸmiÅŸ sÄ±nÄ±flandÄ±rma
rfm_with_country['IleriSeviyeSegment'] = rfm_with_country.apply(
    lambda row: musteri_siniflandir_ileri_seviye(
        row['Recency'], 
        row['Frequency'], 
        row['Monetary'],
        row['Country']
    ),
    axis=1
)

print("\n" + "="*100)
print("ğŸ¯ GELÄ°ÅMÄ°Å MÃœÅTERÄ° SINIFLANDIRMASI - NESTED IF ANALÄ°ZÄ°")
print("="*100)

# Segment analizi
ileri_segment_analiz = rfm_with_country.groupby('IleriSeviyeSegment').agg({
    'CustomerID': 'count',
    'Monetary': 'sum',
    'Frequency': 'mean',
    'Recency': 'mean'
}).round(2)

ileri_segment_analiz.columns = ['MÃ¼ÅŸteri', 'Toplam Gelir (Â£)', 'Ort. Frequency', 'Ort. Recency']
ileri_segment_analiz = ileri_segment_analiz.sort_values('Toplam Gelir (Â£)', ascending=False)

print(ileri_segment_analiz)
print("="*100)

# En deÄŸerli segmentleri vurgulama
print("\nğŸ† EN DEÄERLI 5 SEGMENT:")
top_5 = ileri_segment_analiz.head(5)
for idx, (segment, row) in enumerate(top_5.iterrows(), 1):
    print(f"{idx}. {segment}")
    print(f"   â€¢ MÃ¼ÅŸteri SayÄ±sÄ±: {int(row['MÃ¼ÅŸteri'])}")
    print(f"   â€¢ Toplam Gelir: Â£{row['Toplam Gelir (Â£)']:,.2f}")
    print(f"   â€¢ Gelir PayÄ±: %{(row['Toplam Gelir (Â£)'] / ileri_segment_analiz['Toplam Gelir (Â£)'].sum() * 100):.1f}")
    print()
````

````python
# ğŸ¯ Dinamik FiyatlandÄ±rma Motoru - Complex Nested IF ile AkÄ±llÄ± Pricing
# ==========================================================================

def dinamik_fiyat_hesapla(temel_fiyat, musteri_segment, siparis_miktari, mevsim, gun_tipi):
    """
    Ã‡oklu faktÃ¶rlere gÃ¶re dinamik fiyat hesaplayan geliÅŸmiÅŸ bir fonksiyon.
    YBS'de e-ticaret platformlarÄ±nÄ±n fiyatlandÄ±rma motorlarÄ± bÃ¶yle Ã§alÄ±ÅŸÄ±r.
    
    FaktÃ¶rler:
    1. MÃ¼ÅŸteri Segmenti (sadakat indirimi)
    2. SipariÅŸ MiktarÄ± (bulk indirim)
    3. Mevsim (sezonsal fiyatlandÄ±rma)
    4. GÃ¼n Tipi (hafta sonu vs hafta iÃ§i)
    """
    
    # BaÅŸlangÄ±Ã§ fiyatÄ±
    final_fiyat = temel_fiyat
    indirim_detayi = []
    
    # 1. KATMAN: MÃ¼ÅŸteri Segmenti BazlÄ± Ä°ndirim
    if "VIP Elite" in musteri_segment or "VIP Gold" in musteri_segment:
        musteri_indirim = 0.20  # %20 indirim
        indirim_detayi.append("VIP Ä°ndirimi: %20")
    elif "VIP" in musteri_segment or "Premium" in musteri_segment:
        musteri_indirim = 0.15  # %15 indirim
        indirim_detayi.append("Premium Ä°ndirimi: %15")
    elif "Loyal" in musteri_segment or "Regular" in musteri_segment:
        musteri_indirim = 0.10  # %10 indirim
        indirim_detayi.append("Sadakat Ä°ndirimi: %10")
    else:
        musteri_indirim = 0.05  # %5 temel indirim
        indirim_detayi.append("Standart Ä°ndirim: %5")
    
    # 2. KATMAN: Miktar BazlÄ± Ä°ndirim (Kademeli)
    if siparis_miktari >= 100:
        if siparis_miktari >= 500:
            miktar_indirim = 0.15  # %15 bulk indirim
            indirim_detayi.append("SÃ¼per Bulk Ä°ndirim: %15")
        else:
            miktar_indirim = 0.10  # %10 bulk indirim
            indirim_detayi.append("Bulk Ä°ndirim: %10")
    elif siparis_miktari >= 50:
        miktar_indirim = 0.05  # %5 toplu indirim
        indirim_detayi.append("Toplu Ä°ndirim: %5")
    else:
        miktar_indirim = 0  # Ä°ndirim yok
    
    # 3. KATMAN: Mevsimsel FiyatlandÄ±rma
    if mevsim == "â„ï¸ KÄ±ÅŸ":
        # KÄ±ÅŸ aylarÄ±nda Ã¶zel indirim
        if "UK" in musteri_segment:
            mevsim_carpan = 0.90  # UK iÃ§in %10 indirim
            indirim_detayi.append("KÄ±ÅŸ KampanyasÄ± (UK): %10")
        else:
            mevsim_carpan = 0.95  # DiÄŸerleri iÃ§in %5 indirim
            indirim_detayi.append("KÄ±ÅŸ KampanyasÄ±: %5")
    elif mevsim == "â˜€ï¸ Yaz":
        # Yaz aylarÄ±nda fiyat artÄ±ÅŸÄ± (yÃ¼ksek talep)
        mevsim_carpan = 1.05  # %5 artÄ±ÅŸ
        indirim_detayi.append("Yaz Sezonu: +%5")
    else:
        mevsim_carpan = 1.0  # Normal fiyat
    
    # 4. KATMAN: GÃ¼n Tipi BazlÄ± FiyatlandÄ±rma
    if gun_tipi == "ğŸ‰ Hafta Sonu":
        # Hafta sonu kampanyasÄ±
        gun_carpan = 0.98  # %2 indirim
        indirim_detayi.append("Hafta Sonu FÄ±rsatÄ±: %2")
    else:
        gun_carpan = 1.0  # Normal
    
    # FÄ°YAT HESAPLAMA
    # 1. MÃ¼ÅŸteri ve miktar indirimlerini uygula
    toplam_indirim_orani = musteri_indirim + miktar_indirim
    final_fiyat = final_fiyat * (1 - toplam_indirim_orani)
    
    # 2. Mevsimsel ve gÃ¼n bazlÄ± Ã§arpanlarÄ± uygula
    final_fiyat = final_fiyat * mevsim_carpan * gun_carpan
    
    # 3. Toplam indirim yÃ¼zdesini hesapla
    toplam_indirim = ((temel_fiyat - final_fiyat) / temel_fiyat) * 100
    
    return final_fiyat, toplam_indirim, indirim_detayi


# Ã–rnek senaryolar
print("\n" + "="*110)
print("ğŸ¯ DÄ°NAMÄ°K FÄ°YATLANDIRMA MOTORU - Ã‡OKLU FAKTÃ–R ANALÄ°ZÄ°")
print("="*110)

senaryolar = [
    ("â­â­â­ UK VIP Elite", 500, "â„ï¸ KÄ±ÅŸ", "ğŸ’¼ Hafta Ä°Ã§i"),
    ("ğŸ’ UK Premium Loyal", 100, "â˜€ï¸ Yaz", "ğŸ‰ Hafta Sonu"),
    ("âœ… UK Standard Active", 50, "ğŸŒ¸ Ä°lkbahar", "ğŸ’¼ Hafta Ä°Ã§i"),
    ("ğŸŒ International VIP", 200, "ğŸ‚ Sonbahar", "ğŸ‰ Hafta Sonu"),
    ("ğŸŒ International Entry", 10, "â„ï¸ KÄ±ÅŸ", "ğŸ’¼ Hafta Ä°Ã§i")
]

temel_fiyat = 1000  # Temel Ã¼rÃ¼n fiyatÄ± Â£1000

print(f"\nğŸ“Œ Temel ÃœrÃ¼n FiyatÄ±: Â£{temel_fiyat}\n")

for i, (segment, miktar, mevsim, gun) in enumerate(senaryolar, 1):
    final, indirim_yuzde, detaylar = dinamik_fiyat_hesapla(temel_fiyat, segment, miktar, mevsim, gun)
    
    print(f"{i}. SENARYO:")
    print(f"   ğŸ‘¤ MÃ¼ÅŸteri: {segment}")
    print(f"   ğŸ“¦ Miktar: {miktar} adet")
    print(f"   ğŸŒ Mevsim: {mevsim}")
    print(f"   ğŸ“… GÃ¼n: {gun}")
    print(f"   ğŸ’° Final Fiyat: Â£{final:.2f}")
    print(f"   ğŸ Toplam Ä°ndirim: %{indirim_yuzde:.1f}")
    print(f"   ğŸ“‹ Ä°ndirim DetaylarÄ±:")
    for detay in detaylar:
        print(f"      â€¢ {detay}")
    print("-" * 110)
````
