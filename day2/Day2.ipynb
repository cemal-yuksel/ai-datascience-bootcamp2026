## ğŸ”ª Ã–zel AyÄ±rÄ±cÄ± (Delimiter) KullanÄ±mÄ±

## ğŸ”¢ maxsplit Parametresi

## ğŸ”— join() Metodu - Liste'yi BirleÅŸtirme

    <h4 style="color: #2e7d32; margin-top: 0;">ğŸ“– TanÄ±m</h4>
    <p style="color: #1b5e20; font-size: 1.15em; line-height: 1.8; margin: 0;">
        <code>join()</code> = Liste elemanlarÄ±nÄ± bir <strong>ayÄ±rÄ±cÄ± ile birleÅŸtirerek</strong> tek bir string oluÅŸturur<br><br>

## ğŸ§¹ strip() Ailesi - BoÅŸluk Temizleme

## âœ”ï¸ startswith() ve endswith() - Kontrol MetotlarÄ±

<div style="background-color: #f3e5f5; border: 2px solid #9c27b0; padding: 15px; border-radius: 8px; margin: 15px 0;">
    <h4 style="color: #6a1b9a; margin-top: 0;">ğŸ” BaÅŸlangÄ±Ã§/BitiÅŸ KontrolÃ¼</h4>
    <p style="color: #4a148c; font-size: 1.1em; margin: 0;">
        <code>startswith()</code> = String belirli bir karakterle <strong>baÅŸlÄ±yor mu?</strong><br>
        <code>endswith()</code> = String belirli bir karakterle <strong>bitiyor mu?</strong><br><br>
        ğŸ“¤ <strong>DÃ¶nÃ¼ÅŸ:</strong> True/False
    </p>
</div>

````python
# startswith()
filename = "rapor_2025.pdf"

print("'rapor' ile baÅŸlÄ±yor mu?", filename.startswith("rapor"))    # True
print("'data' ile baÅŸlÄ±yor mu?", filename.startswith("data"))      # False
````

````python
# endswith()
filename = "rapor_2025.pdf"

print(".pdf ile bitiyor mu?", filename.endswith(".pdf"))     # True
print(".xlsx ile bitiyor mu?", filename.endswith(".xlsx"))   # False
````

````python
# Dosya tÃ¼rÃ¼ kontrolÃ¼
files = ["rapor.pdf", "data.xlsx", "kod.py", "image.jpg"]

for file in files:
    if file.endswith(".pdf"):
        print(f"ğŸ“„ {file} - PDF dosyasÄ±")
    elif file.endswith(".xlsx"):
        print(f"ğŸ“Š {file} - Excel dosyasÄ±")
    elif file.endswith(".py"):
        print(f"ğŸ {file} - Python dosyasÄ±")
    else:
        print(f"ğŸ“ {file} - DiÄŸer")
````

    <h4 style="color: #4a148c; margin-top: 0;">ğŸ“‹ Proje AÃ§Ä±klamasÄ±</h4>
    <p style="color: #6a1b9a; font-size: 1.05em; margin: 0;">
        Åifre gÃ¼venliÄŸini kontrol eden bir validasyon sistemi.<br>
        <strong>KullanÄ±lacak Kavramlar:</strong> len(), in, upper(), lower(), isdigit()
    </p>

````python
# Mini Proje 3: Åifre GÃ¼venlik KontrolÃ¼

def check_password(password):
    """Åifre gÃ¼venliÄŸini kontrol et"""
    
    print("=" * 50)
    print("ğŸ” ÅÄ°FRE GÃœVENLÄ°K ANALÄ°ZÄ°")
    print("=" * 50)
    print(f"\nTest edilen ÅŸifre: {'*' * len(password)}")
    print(f"Uzunluk: {len(password)} karakter\n")
    
    # GÃ¼venlik kriterleri
    score = 0
    feedback = []
    
    # 1. Uzunluk kontrolÃ¼
    if len(password) >= 8:
        print("âœ… Uzunluk: 8+ karakter (GÃ¼Ã§lÃ¼)")
        score += 25
    elif len(password) >= 6:
        print("âš ï¸ Uzunluk: 6-7 karakter (Orta)")
        score += 15
        feedback.append("Daha uzun ÅŸifre kullanÄ±n")
    else:
        print("âŒ Uzunluk: 6'dan az (ZayÄ±f)")
        feedback.append("En az 8 karakter kullanÄ±n")
    
    # 2. BÃ¼yÃ¼k harf kontrolÃ¼
    has_upper = False
    for char in password:
        if char.isupper():
            has_upper = True
            break
    
    if has_upper:
        print("âœ… BÃ¼yÃ¼k harf iÃ§eriyor")
        score += 25
    else:
        print("âŒ BÃ¼yÃ¼k harf iÃ§ermiyor")
        feedback.append("En az 1 bÃ¼yÃ¼k harf ekleyin")
    
    # 3. KÃ¼Ã§Ã¼k harf kontrolÃ¼
    has_lower = False
    for char in password:
        if char.islower():
            has_lower = True
            break
    
    if has_lower:
        print("âœ… KÃ¼Ã§Ã¼k harf iÃ§eriyor")
        score += 25
    else:
        print("âŒ KÃ¼Ã§Ã¼k harf iÃ§ermiyor")
        feedback.append("En az 1 kÃ¼Ã§Ã¼k harf ekleyin")
    
    # 4. Rakam kontrolÃ¼
    has_digit = False
    for char in password:
        if char.isdigit():
            has_digit = True
            break
    
    if has_digit:
        print("âœ… Rakam iÃ§eriyor")
        score += 25
    else:
        print("âŒ Rakam iÃ§ermiyor")
        feedback.append("En az 1 rakam ekleyin")
    
    # GÃ¼venlik skoru
    print("\n" + "=" * 50)
    print(f"ğŸ“Š GÃœVENLÄ°K SKORU: {score}/100")
    
    if score == 100:
        print("ğŸŒŸ MÃ¼kemmel! Åifreniz Ã§ok gÃ¼Ã§lÃ¼.")
    elif score >= 75:
        print("âœ… Ä°yi! Åifreniz gÃ¼venli.")
    elif score >= 50:
        print("âš ï¸ Orta! Åifrenizi gÃ¼Ã§lendirebilirsiniz.")
    else:
        print("âŒ ZayÄ±f! Åifrenizi deÄŸiÅŸtirmeniz Ã¶nerilir.")
    
    # Ã–neriler
    if feedback:
        print("\nğŸ’¡ Ã–NERÄ°LER:")
        for i, suggestion in enumerate(feedback, 1):
            print(f"  {i}. {suggestion}")
    
    print("=" * 50)

# Test senaryolarÄ±
print("TEST 1: ZayÄ±f Åifre")
check_password("abc123")

print("\n\nTEST 2: Orta Åifre")
check_password("Password")

print("\n\nTEST 3: GÃ¼Ã§lÃ¼ Åifre")
check_password("MyP@ssw0rd123")
````
