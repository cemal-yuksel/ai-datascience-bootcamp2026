## ğŸ“ String'ler ve DeÄŸiÅŸtirilemezlik

String'ler Python'da deÄŸiÅŸtirilemez (immutable) veri yapÄ±larÄ±dÄ±r. Bir string'in iÃ§eriÄŸini deÄŸiÅŸtiremeyiz, sadece yeni bir string oluÅŸturabiliriz.

````python
# String Ã¶rneÄŸi
myString = "hello python"
print("Original String:", myString)
print("Type:", type(myString))

# String'den bir karaktere eriÅŸim
print("\nÄ°lk karakter:", myString[0])

# String'i deÄŸiÅŸtirmeye Ã§alÄ±ÅŸalÄ±m (HATA VERECEK!)
# myString[0] = "H"  # TypeError: 'str' object does not support item assignment

# String'i deÄŸiÅŸtirmek iÃ§in yeni bir string oluÅŸturmalÄ±yÄ±z
myString = "Hello Python"
print("\nYeni String:", myString)
````

## ğŸ”¨ String'i List'e DÃ¶nÃ¼ÅŸtÃ¼rme - split() Metodu

`split()` metodu bir string'i boÅŸluklara (veya belirtilen karaktere) gÃ¶re bÃ¶ler ve bir **list** dÃ¶ndÃ¼rÃ¼r.

````python
# String'i list'e dÃ¶nÃ¼ÅŸtÃ¼rme
myString = "hello python"
wordList = myString.split()

print("String:", myString)
print("List:", wordList)
print("List'in tipi:", type(wordList))
print("List'in tipi:", type(myString.split()))

# MIS Ã–rneÄŸi: Åirket departmanlarÄ±
departments = "Ä°nsan KaynaklarÄ± Finans Pazarlama Bilgi Teknolojileri"
department_list = departments.split()
print("\nğŸ¢ Departmanlar:", department_list)
````

# 4ï¸âƒ£ Mutability (DeÄŸiÅŸtirilebilirlik) KavramÄ±

<div style="background: linear-gradient(135deg, #a8edea 0%, #fed6e3 100%); padding: 25px; border-radius: 12px; color: #333; margin: 20px 0; box-shadow: 0 8px 20px rgba(0,0,0,0.2);">
    <h2 style="margin: 0; font-size: 2em;">âœï¸ List'ler DeÄŸiÅŸtirilebilir!</h2>
    <p style="font-size: 1.2em; margin-top: 15px;">String'lerden farklÄ± olarak, list elemanlarÄ±nÄ± doÄŸrudan deÄŸiÅŸtirebiliriz!</p>

# 5ï¸âƒ£ Negatif Ä°ndeksleme

````python
myList = [100, 20, 30, 40, 50, 60, 70]

print("List:", myList)
print("Uzunluk:", len(myList))

# Pozitif indeks ile son eleman
print("\nSon eleman (pozitif):", myList[len(myList)-1])

# Negatif indeks ile son eleman (DAHA KOLAY!)
print("Son eleman (negatif):", myList[-1])
print("Sondan ikinci:", myList[-2])
print("Sondan Ã¼Ã§Ã¼ncÃ¼:", myList[-3])

# MIS Ã–rneÄŸi: En son girilen kayÄ±tlar
transactions = [1500, 2300, 1800, 900, 3500, 2100]
print("\nğŸ’° Son iÅŸlem:", transactions[-1], "TL")
print("Sondan ikinci iÅŸlem:", transactions[-2], "TL")
print("Son 3 iÅŸlem toplamÄ±:", sum(transactions[-3:]), "TL")
````

# 7ï¸âƒ£ List MetodlarÄ± - Ã‡Ä±karma ve DÃ¼zenleme Ä°ÅŸlemleri

<div style="background: linear-gradient(135deg, #ff6b6b 0%, #feca57 100%); padding: 25px; border-radius: 12px; color: white; margin: 20px 0; box-shadow: 0 8px 20px rgba(0,0,0,0.2);">
    <h2 style="margin: 0; font-size: 2em;">ğŸ—‘ï¸ Silme, SÄ±ralama ve Arama MetodlarÄ±</h2>

## ğŸ”¹ count() ve index() - Arama MetodlarÄ±

````python
myList = [100, 20, 35, 30, 40, 50, 60, 70, 80, 1]

# count() - bir elemanÄ±n listede kaÃ§ kez geÃ§tiÄŸini sayar
print("List:", myList)
print("20'nin sayÄ±sÄ±:", myList.count(20))
print("25'in sayÄ±sÄ±:", myList.count(25))

# index() - elemanÄ±n ilk bulunduÄŸu indeksi dÃ¶ndÃ¼rÃ¼r
print("\n40'Ä±n indeksi:", myList.index(40))

# MIS Ã–rneÄŸi
products = ["Laptop", "Mouse", "Keyboard", "Mouse", "Monitor", "Mouse"]
print("\nğŸ–±ï¸ Mouse sayÄ±sÄ±:", products.count("Mouse"))
print("Keyboard'Ä±n pozisyonu:", products.index("Keyboard"))
````

## ğŸ”¹ pop() ve remove() - Silme MetodlarÄ±

````python
myList = [100, 20, 35, 30, 40, 50, 60, 70, 80, 1]
print("BaÅŸlangÄ±Ã§:", myList)

# pop() - son elemanÄ± Ã§Ä±karÄ±r ve dÃ¶ndÃ¼rÃ¼r
removed_item = myList.pop()
print(f"\nÃ‡Ä±karÄ±lan eleman: {removed_item}")
print("Pop sonrasÄ± list:", myList)

# remove() - belirtilen deÄŸeri Ã§Ä±karÄ±r
myList.remove(40)
print("\n40 silindi:", myList)

# MIS Ã–rneÄŸi: Stoktan Ã¼rÃ¼n Ã§Ä±karma
inventory = ["Laptop", "Mouse", "Keyboard", "Monitor", "Headset"]
print("\nğŸ“¦ Envanter:", inventory)

# Son Ã¼rÃ¼n satÄ±ldÄ±
sold = inventory.pop()
print(f"SatÄ±lan Ã¼rÃ¼n: {sold}")
print("GÃ¼ncel envanter:", inventory)

# Belirli bir Ã¼rÃ¼n kaldÄ±rÄ±ldÄ±
inventory.remove("Mouse")
print("Mouse kaldÄ±rÄ±ldÄ±:", inventory)
````

## ğŸ”¹ reverse() ve sort() - DÃ¼zenleme MetodlarÄ±

````python
myList = [80, 70, 60, 50, 30, 35, 20, 100]
print("Orijinal:", myList)

# reverse() - listeyi ters Ã§evirir
myList.reverse()
print("\nTers Ã§evrilmiÅŸ:", myList)

# sort() - listeyi kÃ¼Ã§Ã¼kten bÃ¼yÃ¼ÄŸe sÄ±ralar
myList.sort()
print("SÄ±ralanmÄ±ÅŸ:", myList)

# MIS Ã–rneÄŸi: Performans skorlarÄ±
employees = ["Ali", "AyÅŸe", "Mehmet", "Zeynep", "Can"]
scores = [85, 92, 78, 95, 88]

print("\nğŸ“Š Ã‡alÄ±ÅŸan PerformanslarÄ±:")
for i in range(len(employees)):
    print(f"{employees[i]}: {scores[i]}")

# SkorlarÄ± sÄ±ralayalÄ±m
scores.sort()
print("\nSÄ±ralanmÄ±ÅŸ skorlar:", scores)

# BÃ¼yÃ¼kten kÃ¼Ã§Ã¼ÄŸe sÄ±ralamak iÃ§in
scores.sort(reverse=True)
print("Azalan sÄ±rada:", scores)
````

````python
# Nested list oluÅŸturma
myNestedList = [10, 20, 3.14, "atil", [1, 2, 3]]

print("Nested List:", myNestedList)
print("\nEleman 0:", myNestedList[0])
print("Eleman 1:", myNestedList[1])
print("Eleman 2:", myNestedList[2])
print("Eleman 3:", myNestedList[3])
print("Eleman 4:", myNestedList[4])

# Ä°Ã§ list'e eriÅŸim
innerList = myNestedList[4]
print("\nÄ°Ã§ List:", innerList)
print("Tip:", type(innerList))

# Negatif indeks ile
innerList2 = myNestedList[-1]
print("\nNegatif indeks ile:", innerList2)
print("Ä°Ã§ listin 2. elemanÄ±:", innerList2[1])

# DoÄŸrudan eriÅŸim
print("\nDoÄŸrudan eriÅŸim:", myNestedList[4][2])

# Daha karmaÅŸÄ±k Ã¶rnek
lastList = ["a", "b", ["c", "d", "e"], "f"]
print("\n\nKarmaÅŸÄ±k List:", lastList)
print("Uzunluk:", len(lastList))
print("Ä°Ã§ listin 2. elemanÄ±:", lastList[2][1])

# MIS Ã–rneÄŸi: Departman bazlÄ± Ã§alÄ±ÅŸan listeleri
company = [
    ["Ä°nsan KaynaklarÄ±", ["Ali", "AyÅŸe", "Mehmet"]],
    ["Finans", ["Zeynep", "Can", "Deniz", "Ece"]],
    ["BT", ["Ahmet", "Selin"]]
]

print("\nğŸ¢ Åirket YapÄ±sÄ±:")
for dept in company:
    dept_name = dept[0]
    employees = dept[1]
    print(f"{dept_name}: {len(employees)} Ã§alÄ±ÅŸan - {employees}")
````

# 1ï¸âƒ£2ï¸âƒ£ List Slicing (Dilimleme) Ä°ÅŸlemleri

<div style="background: linear-gradient(135deg, #ffecd2 0%, #fcb69f 100%); padding: 25px; border-radius: 12px; color: #333; margin: 20px 0; box-shadow: 0 8px 20px rgba(0,0,0,0.2);">
    <h2 style="margin: 0; font-size: 2em;">âœ‚ï¸ List Dilimleme - Slicing</h2>
    <p style="font-size: 1.2em; margin-top: 15px;">List'in belirli bÃ¶lÃ¼mlerini almak iÃ§in slicing kullanÄ±yoruz!</p>
    <p style="font-size: 1.1em; margin-top: 10px;"><strong>Syntax:</strong> <code>list[start:stop:step]</code></p>
</div>

````python
firstList = [10, 20, 30, 40, 50, 60, 70]
print("Orijinal List:", firstList)

# TÃ¼m listeyi alma
print("\nfirstList[::]  â†’", firstList[::])

# Belirli bir indeksten baÅŸlama
print("firstList[2::] â†’", firstList[2::])

# Belirli bir indekse kadar alma
print("firstList[:5:] â†’", firstList[:5:])

# AdÄ±m (step) ile alma
print("firstList[::2] â†’", firstList[::2])

# BaÅŸlangÄ±Ã§, bitiÅŸ ve adÄ±m birlikte
print("firstList[1:6:2] â†’", firstList[1:6:2])

# MIS Ã–rneÄŸi: AylÄ±k verilerden Ã§eyreklik analiz
monthly_sales = [100, 120, 110, 150, 160, 155, 180, 190, 170, 200, 210, 220]
months = ["Oca", "Åub", "Mar", "Nis", "May", "Haz", "Tem", "AÄŸu", "Eyl", "Eki", "Kas", "Ara"]

print("\nğŸ“Š AylÄ±k SatÄ±ÅŸlar:", monthly_sales)

# Ã‡eyreklik veriler
q1 = monthly_sales[0:3]  # Ä°lk 3 ay
q2 = monthly_sales[3:6]  # Ä°kinci 3 ay
q3 = monthly_sales[6:9]  # ÃœÃ§Ã¼ncÃ¼ 3 ay
q4 = monthly_sales[9:12] # Son 3 ay

print(f"\nÃ‡eyrek 1 ({months[0:3]}): {q1} - Toplam: {sum(q1)}")
print(f"Ã‡eyrek 2 ({months[3:6]}): {q2} - Toplam: {sum(q2)}")
print(f"Ã‡eyrek 3 ({months[6:9]}): {q3} - Toplam: {sum(q3)}")
print(f"Ã‡eyrek 4 ({months[9:12]}): {q4} - Toplam: {sum(q4)}")

# Tek aylar
odd_months = monthly_sales[::2]
print(f"\nTek aylÄ±k veriler: {odd_months}")

# Son 3 ay
last_quarter = monthly_sales[-3:]
print(f"Son Ã§eyrek: {last_quarter}")
````
