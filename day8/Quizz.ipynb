```python
# ============================================================================
# CEVAP 1: MANTIKSAL OPERATÃ–RLER - AND (VE) OPERATÃ–RÃœ
# ============================================================================

# ğŸ“Œ DEÄERLER:
x = 5
y = 3
z = 6

# ğŸ“Œ Ä°FADE: x > y and z > x
# Bu ifadeyi adÄ±m adÄ±m analiz edelim:

# ADIM 1: Ä°lk koÅŸulu kontrol et (x > y)
print("="*60)
print("ğŸ” MANTIKSAL OPERATÃ–R ANALÄ°ZÄ° - AND (VE)")
print("="*60)
print(f"\nğŸ“Š DeÄŸerler: x={x}, y={y}, z={z}")
print("\nğŸ“ Ä°fade: x > y and z > x")
print("\n--- ADIM 1: Ä°lk KoÅŸul ---")
print(f"x > y  âœ  {x} > {y}  âœ  {x > y}")

# ADIM 2: Ä°kinci koÅŸulu kontrol et (z > x)
print("\n--- ADIM 2: Ä°kinci KoÅŸul ---")
print(f"z > x  âœ  {z} > {x}  âœ  {z > x}")

# ADIM 3: AND operatÃ¶rÃ¼ ile birleÅŸtir
print("\n--- ADIM 3: AND OperatÃ¶rÃ¼ ile BirleÅŸtirme ---")
print(f"{x > y} AND {z > x}  âœ  {x > y and z > x}")

# ğŸ¯ SONUÃ‡
sonuc = x > y and z > x
print("\n" + "="*60)
print(f"âœ… SONUÃ‡: {sonuc}")
print("="*60)

# ğŸ“š BÄ°LGÄ° NOTU:
# AND (ve) operatÃ¶rÃ¼: Her iki koÅŸul da True olmalÄ± ki sonuÃ§ True olsun
# True  AND True  = True
# True  AND False = False
# False AND True  = False
# False AND False = False
```

```python
# 2) AynÄ± deÄŸerlerle kod ÅŸu ÅŸekilde deÄŸiÅŸtilirse Ã§Ä±ktÄ± ne olacaktÄ±r?
```

x < y or z > y

```python
# ============================================================================
# CEVAP 2: MANTIKSAL OPERATÃ–RLER - OR (VEYA) OPERATÃ–RÃœ
# ============================================================================

# ğŸ“Œ DEÄERLER: (AynÄ± deÄŸerler kullanÄ±lÄ±yor)
x = 5
y = 3
z = 6

# ğŸ“Œ Ä°FADE: x < y or z > y
# Bu ifadeyi adÄ±m adÄ±m analiz edelim:

print("="*60)
print("ğŸ” MANTIKSAL OPERATÃ–R ANALÄ°ZÄ° - OR (VEYA)")
print("="*60)
print(f"\nğŸ“Š DeÄŸerler: x={x}, y={y}, z={z}")
print("\nğŸ“ Ä°fade: x < y or z > y")

# ADIM 1: Ä°lk koÅŸulu kontrol et (x < y)
print("\n--- ADIM 1: Ä°lk KoÅŸul ---")
print(f"x < y  âœ  {x} < {y}  âœ  {x < y}")
print("ğŸ’¡ 5 kÃ¼Ã§Ã¼k deÄŸil 3'ten, dolayÄ±sÄ±yla False")

# ADIM 2: Ä°kinci koÅŸulu kontrol et (z > y)
print("\n--- ADIM 2: Ä°kinci KoÅŸul ---")
print(f"z > y  âœ  {z} > {y}  âœ  {z > y}")
print("ğŸ’¡ 6 bÃ¼yÃ¼k 3'ten, dolayÄ±sÄ±yla True")

# ADIM 3: OR operatÃ¶rÃ¼ ile birleÅŸtir
print("\n--- ADIM 3: OR OperatÃ¶rÃ¼ ile BirleÅŸtirme ---")
print(f"{x < y} OR {z > y}  âœ  {x < y or z > y}")
print("ğŸ’¡ OR operatÃ¶rÃ¼nde en az bir koÅŸul True ise sonuÃ§ True olur!")

# ğŸ¯ SONUÃ‡
sonuc = x < y or z > y
print("\n" + "="*60)
print(f"âœ… SONUÃ‡: {sonuc}")
print("="*60)

# ğŸ“š BÄ°LGÄ° NOTU:
# OR (veya) operatÃ¶rÃ¼: En az bir koÅŸul True ise sonuÃ§ True olur
# True  OR True  = True
# True  OR False = True
# False OR True  = True
# False OR False = False

# ğŸ”¥ KARÅILAÅTIRMA: AND vs OR
print("\n" + "ğŸ”¥ AND vs OR KARÅILAÅTIRMASI" + "\n")
print("AND: Her iki koÅŸul da True olmalÄ± âœ Daha katÄ±")
print("OR:  En az bir koÅŸul True olmalÄ± âœ Daha esnek")
```

```python
# ============================================================================
# ALTERNATÄ°F Ã‡Ã–ZÃœMLER - Dictionary DeÄŸer KontrolÃ¼
# ============================================================================

print("\n" + "="*70)
print("ğŸ”§ YÃ–NTEM 2: For DÃ¶ngÃ¼sÃ¼ ile Manuel Kontrol")
print("="*70)

my_dictionary = {"k1":10, "k2k":"a", "k32":30, "k4":"c"}

# Boolean deÄŸiÅŸken ile kontrol
bulundu = False

print("\nğŸ” DeÄŸerleri tek tek kontrol ediyoruz:")
for anahtar, deger in my_dictionary.items():
    print(f"   '{anahtar}' â†’ {deger} ", end="")
    if deger == "c":
        print("âœ“ [BULUNDU!]")
        bulundu = True
    else:
        print("âœ—")

if bulundu:
    print("\nâœ… SONUÃ‡: 'c' harfi deÄŸerler iÃ§inde bulunuyor!")
else:
    print("\nâŒ SONUÃ‡: 'c' harfi deÄŸerler iÃ§inde bulunmuyor!")

# ============================================================================
print("\n" + "="*70)
print("ğŸ”§ YÃ–NTEM 3: List Comprehension ile Kontrol")
print("="*70)

# TÃ¼m deÄŸerleri liste olarak al ve kontrol et
deger_listesi = [deger for deger in my_dictionary.values()]
print(f"\nDeÄŸer listesi: {deger_listesi}")

if "c" in deger_listesi:
    print("âœ… SONUÃ‡: 'c' harfi deÄŸerler iÃ§inde bulunuyor!")
else:
    print("âŒ SONUÃ‡: 'c' harfi deÄŸerler iÃ§inde bulunmuyor!")

# ============================================================================
print("\n" + "="*70)
print("ğŸ“š Ã–NEMLÄ° NOTLAR VE BEST PRACTICES")
print("="*70)
print("""
ğŸ”‘ Dictionary Temel MetodlarÄ±:
   â€¢ .keys()    â†’ TÃ¼m anahtarlarÄ± dÃ¶ndÃ¼rÃ¼r
   â€¢ .values()  â†’ TÃ¼m deÄŸerleri dÃ¶ndÃ¼rÃ¼r
   â€¢ .items()   â†’ TÃ¼m anahtar-deÄŸer Ã§iftlerini dÃ¶ndÃ¼rÃ¼r

ğŸ’¡ En Ä°yi YÃ¶ntem (Best Practice):
   â€¢ Basit kontroller iÃ§in: "x" in dict.values()
   â€¢ DetaylÄ± bilgi iÃ§in: for dÃ¶ngÃ¼sÃ¼ ile .items()
   â€¢ Performans kritikse: Set dÃ¶nÃ¼ÅŸÃ¼mÃ¼ ile arama

âš ï¸  Dikkat Edilmesi Gerekenler:
   â€¢ Dictionary deÄŸerleri farklÄ± tiplerde olabilir (int, str, list, vb.)
   â€¢ Tip kontrolÃ¼ yaparken type() fonksiyonunu kullanÄ±n
   â€¢ String kontrolÃ¼ yapÄ±lÄ±yorsa bÃ¼yÃ¼k/kÃ¼Ã§Ã¼k harf duyarlÄ±lÄ±ÄŸÄ±na dikkat!

ğŸ¯ Bu Ã–rnekte:
   â€¢ SÃ¶zlÃ¼ÄŸÃ¼mÃ¼z: {"k1":10, "k2k":"a", "k32":30, "k4":"c"}
   â€¢ AradÄ±ÄŸÄ±mÄ±z: "c" karakteri
   â€¢ BulunduÄŸu yer: "k4" anahtarÄ±nÄ±n deÄŸeri
   â€¢ SonuÃ§: âœ… BULUNDU!
""")
```

```python
#5) AÅŸaÄŸÄ±daki sÃ¶zlÃ¼kte, anahtarlar iÃ§inde a harfinin geÃ§ip geÃ§mediÄŸini gÃ¶steren bir if koÅŸulu yazÄ±nÄ±z
```

```python
my_other_dictionary = {"b":203,"c":"a","a":400,"d":"f"}
```

```python
# ============================================================================
# CEVAP 5: DICTIONARY (SÃ–ZLÃœK) - ANAHTARLARDA ELEMAN ARAMA
# ============================================================================

# ğŸ“Œ SÃ–ZLÃœK YAPISI:
my_other_dictionary = {"b":203, "c":"a", "a":400, "d":"f"}

print("="*70)
print("ğŸ”‘ DICTIONARY (SÃ–ZLÃœK) - ANAHTARLARDA ELEMAN KONTROLÃœ")
print("="*70)

# ğŸ” SÃ–ZLÃœÄÃœMÃœZÃœ Ä°NCELEYELÄ°M
print("\nğŸ” SÃ¶zlÃ¼ÄŸÃ¼mÃ¼z:")
print(f"{my_other_dictionary}\n")

print("ğŸ“Š Anahtar-DeÄŸer Ã‡iftleri:")
for anahtar, deger in my_other_dictionary.items():
    print(f"   Anahtar: '{anahtar}' (Tip: {type(anahtar).__name__}) â†’ DeÄŸer: {deger}")

# ğŸ“Œ PROBLEM: Anahtarlar iÃ§inde 'a' harfi var mÄ±?
print("\n" + "="*70)
print("â“ SORU: Anahtarlar iÃ§inde 'a' harfi geÃ§iyor mu?")
print("="*70)

# ğŸ”§ YÃ–NTEM 1: keys() metodu + in operatÃ¶rÃ¼ (EN BASIT VE ETKÄ°LÄ°)
print("\nğŸ”§ YÃ–NTEM 1: .keys() Metodu ile Kontrol")
print("-" * 70)

# Dictionary'nin tÃ¼m anahtarlarÄ±nÄ± alalÄ±m
anahtarlar = my_other_dictionary.keys()
print(f"TÃ¼m anahtarlar: {list(anahtarlar)}")

# 'a' harfi anahtarlar arasÄ±nda var mÄ± kontrol edelim
if "a" in my_other_dictionary.keys():
    print("\nâœ… SONUÃ‡: Evet, 'a' harfi anahtarlar iÃ§inde bulunuyor!")
    print(f"   ğŸ“ Anahtar: 'a' â†’ DeÄŸer: {my_other_dictionary['a']}")
else:
    print("\nâŒ SONUÃ‡: HayÄ±r, 'a' harfi anahtarlar iÃ§inde bulunmuyor!")

# ğŸ’¡ BONUS BÄ°LGÄ°: Daha kÄ±sa yÃ¶ntem
print("\n" + "="*70)
print("ğŸ’¡ BONUS: Daha KÄ±sa YÃ¶ntem")
print("="*70)
print("\n.keys() metodunu kullanmadan da kontrol edebiliriz:")
print("Ã‡Ã¼nkÃ¼ Python'da 'in' operatÃ¶rÃ¼ dictionary'lerde varsayÄ±lan olarak")
print("anahtarlarda arama yapar!\n")

# Bu iki yÃ¶ntem aynÄ± sonucu verir:
print(f"'a' in my_other_dictionary.keys()  â†’  {'a' in my_other_dictionary.keys()}")
print(f"'a' in my_other_dictionary         â†’  {'a' in my_other_dictionary}")
print("\nâš¡ Ä°kisi de aynÄ±! KÄ±sa olanÄ± tercih edilir (Pythonic kod)")
```

```python
# ============================================================================
# ALTERNATÄ°F Ã‡Ã–ZÃœMLER - Dictionary Anahtar KontrolÃ¼
# ============================================================================

print("\n" + "="*70)
print("ğŸ”§ YÃ–NTEM 2: For DÃ¶ngÃ¼sÃ¼ ile Manuel Kontrol")
print("="*70)

my_other_dictionary = {"b":203, "c":"a", "a":400, "d":"f"}

# Boolean deÄŸiÅŸken ile kontrol
bulundu = False
bulunan_anahtar = None

print("\nğŸ” AnahtarlarÄ± tek tek kontrol ediyoruz:")
for anahtar in my_other_dictionary.keys():
    print(f"   Anahtar: '{anahtar}' ", end="")
    if anahtar == "a":
        print("âœ“ [BULUNDU!]")
        bulundu = True
        bulunan_anahtar = anahtar
    else:
        print("âœ—")

if bulundu:
    print(f"\nâœ… SONUÃ‡: 'a' harfi anahtarlar iÃ§inde bulunuyor!")
    print(f"   ğŸ“ DeÄŸeri: {my_other_dictionary[bulunan_anahtar]}")
else:
    print("\nâŒ SONUÃ‡: 'a' harfi anahtarlar iÃ§inde bulunmuyor!")

# ============================================================================
print("\n" + "="*70)
print("ğŸ”§ YÃ–NTEM 3: get() Metodu ile GÃ¼venli EriÅŸim")
print("="*70)

print("\nğŸ’¡ get() metodu: Anahtar yoksa hata vermez, None dÃ¶ndÃ¼rÃ¼r")
print("-" * 70)

# 'a' anahtarÄ± var mÄ± kontrol et
deger = my_other_dictionary.get("a")

if deger is not None:
    print(f"\nâœ… 'a' anahtarÄ± bulundu!")
    print(f"   DeÄŸeri: {deger}")
else:
    print("\nâŒ 'a' anahtarÄ± bulunamadÄ±!")

# Olmayan bir anahtar deneyelim
print("\nğŸ” Olmayan bir anahtar ('z') iÃ§in test:")
deger_z = my_other_dictionary.get("z")
print(f"   my_other_dictionary.get('z')  â†’  {deger_z}")
print("   ğŸ’¡ KeyError hatasÄ± vermiyor, None dÃ¶ndÃ¼rÃ¼yor!")

# ============================================================================
print("\n" + "="*70)
print("ğŸ“š DEÄER vs ANAHTAR KONTROLÃœ KARÅILAÅTIRMASI")
print("="*70)
print("""
ğŸ”‘ ANAHTAR (KEY) KONTROLÃœ:
   â€¢ "x" in dictionary.keys()  veya  "x" in dictionary
   â€¢ Anahtarlar benzersizdir (unique)
   â€¢ HÄ±zlÄ± eriÅŸim: O(1) time complexity
   â€¢ dictionary["key"] ile deÄŸere eriÅŸim

ğŸ’ DEÄER (VALUE) KONTROLÃœ:
   â€¢ "x" in dictionary.values()
   â€¢ DeÄŸerler tekrarlÄ± olabilir
   â€¢ YavaÅŸ arama: O(n) time complexity
   â€¢ DeÄŸerden anahtara gitmek iÃ§in dÃ¶ngÃ¼ gerekir

ğŸ¯ Bu Ã–rnekte:
   â€¢ SÃ¶zlÃ¼ÄŸÃ¼mÃ¼z: {"b":203, "c":"a", "a":400, "d":"f"}
   â€¢ AradÄ±ÄŸÄ±mÄ±z: "a" harfi (ANAHTARLARDA)
   â€¢ BulunduÄŸu yer: Anahtar olarak mevcut
   â€¢ DeÄŸeri: 400
   â€¢ SonuÃ§: âœ… BULUNDU!

âš¡ PERFORMANS Ä°PUCU:
   â€¢ Anahtar kontrolÃ¼ Ã§ok hÄ±zlÄ±dÄ±r (hash table)
   â€¢ DeÄŸer kontrolÃ¼ daha yavaÅŸtÄ±r (tÃ¼m deÄŸerleri tarar)
   â€¢ SÄ±k arama yapÄ±lacaksa anahtarlarÄ± kullan!
""")
```

```python
# ============================================================================
# CEVAP 6: LÄ°STE FÄ°LTRELEME - Ã‡Ä°FT SAYILARI BULMA
# ============================================================================

# ğŸ“Œ LÄ°STE YAPISI:
my_numbers = [1,2,3,4,5,6,19,20,32,21,20,1111,23,24]

print("="*70)
print("ğŸ”¢ LÄ°STE FÄ°LTRELEME - Ã‡Ä°FT SAYILARI BULMA VE YAZDIRMA")
print("="*70)

# ğŸ“š Ã‡Ä°FT SAYI NEDÄ°R?
print("\nğŸ“š Ã‡ift SayÄ± Nedir?")
print("-" * 70)
print("Ã‡ift sayÄ±: 2'ye tam bÃ¶lÃ¼nebilen sayÄ±lardÄ±r")
print("Matematiksel kontrol: SayÄ± % 2 == 0")
print("% (modÃ¼l) operatÃ¶rÃ¼: BÃ¶lme iÅŸleminden kalanÄ± verir")
print("-" * 70)

# ğŸ” LÄ°STEMÄ°ZÄ° Ä°NCELEYELÄ°M
print(f"\nğŸ” Listemiz: {my_numbers}")
print(f"ğŸ“Š Eleman sayÄ±sÄ±: {len(my_numbers)}")

# ğŸ“Œ PROBLEM: Sadece Ã§ift sayÄ±larÄ± yazdÄ±r
print("\n" + "="*70)
print("â“ SORU: Listedeki sadece Ã‡Ä°FT sayÄ±larÄ± yazdÄ±r")
print("="*70)

# ğŸ”§ YÃ–NTEM 1: For DÃ¶ngÃ¼sÃ¼ + if KoÅŸulu (TEMEL YÃ–NTEM)
print("\nğŸ”§ YÃ–NTEM 1: For DÃ¶ngÃ¼sÃ¼ + if KoÅŸulu")
print("-" * 70)

print("\nğŸ” TÃ¼m sayÄ±larÄ± tek tek kontrol ediyoruz:\n")

cift_sayilar_sayaci = 0

for sayi in my_numbers:
    # Her sayÄ± iÃ§in modÃ¼l kontrolÃ¼ yap
    kalan = sayi % 2
    
    if kalan == 0:  # Ã‡ift sayÄ±
        print(f"   {sayi:4} % 2 = {kalan}  âœ“ Ã‡Ä°FT  â†’ YazdÄ±r!")
        cift_sayilar_sayaci += 1
    else:  # Tek sayÄ±
        print(f"   {sayi:4} % 2 = {kalan}  âœ— TEK")

print("\n" + "-" * 70)
print(f"ğŸ“Š Toplam {cift_sayilar_sayaci} adet Ã§ift sayÄ± bulundu!")
print("-" * 70)

# SONUÃ‡: Sadece Ã§ift sayÄ±larÄ± yazdÄ±r
print("\nğŸ¯ Ã‡IKTI: Sadece Ã‡ift SayÄ±lar")
print("="*70)

for sayi in my_numbers:
    if sayi % 2 == 0:
        print(sayi, end=" ")

print("\n" + "="*70)
```

```python
# ============================================================================
# YÃ–NTEM 2: Ã‡ift SayÄ±larÄ± Listeye Kaydetme
# ============================================================================

my_numbers = [1,2,3,4,5,6,19,20,32,21,20,1111,23,24]

print("\n" + "="*70)
print("ğŸ”§ YÃ–NTEM 2: Ã‡ift SayÄ±larÄ± Yeni Bir Listeye Kaydetme")
print("="*70)

# BoÅŸ bir liste oluÅŸtur
cift_sayilar = []

print("\nğŸ”„ Listeyi tarÄ±yoruz ve Ã§ift sayÄ±larÄ± ekliyoruz...\n")

for sayi in my_numbers:
    if sayi % 2 == 0:
        cift_sayilar.append(sayi)  # Ã‡ift sayÄ±yÄ± listeye ekle
        print(f"   âœ“ {sayi} Ã§ift sayÄ± â†’ Listeye eklendi!")

print("\n" + "-" * 70)
print(f"ğŸ“‹ Ã‡ift SayÄ±lar Listesi: {cift_sayilar}")
print(f"ğŸ“Š Toplam {len(cift_sayilar)} adet Ã§ift sayÄ±")
print("-" * 70)

# Listeyi yazdÄ±r
print("\nğŸ¯ Ã‡IKTI:")
print("="*70)
for sayi in cift_sayilar:
    print(sayi, end=" ")
print("\n" + "="*70)
```

```python
# ============================================================================
# YÃ–NTEM 3: List Comprehension (Pythonic YÃ¶ntem) + Ek Bilgiler
# ============================================================================

my_numbers = [1,2,3,4,5,6,19,20,32,21,20,1111,23,24]

print("\n" + "="*70)
print("ğŸ”§ YÃ–NTEM 3: List Comprehension (Ä°LERÄ° SEVÄ°YE - PYTHONIC)")
print("="*70)

print("\nğŸ’¡ List Comprehension Nedir?")
print("-" * 70)
print("Liste oluÅŸturmanÄ±n kÄ±sa ve okunabilir yolu")
print("SÃ¶zdizimi: [ifade for eleman in liste if koÅŸul]")
print("-" * 70)

# List comprehension ile Ã§ift sayÄ±larÄ± filtrele
cift_sayilar_pythonic = [sayi for sayi in my_numbers if sayi % 2 == 0]

print(f"\nğŸ“ Kod: [sayi for sayi in my_numbers if sayi % 2 == 0]")
print(f"ğŸ“‹ SonuÃ§: {cift_sayilar_pythonic}")

# KarÅŸÄ±laÅŸtÄ±rma: Klasik yÃ¶ntem vs List Comprehension
print("\n" + "="*70)
print("âš–ï¸  YÃ–NTEMLER KARÅILAÅTIRMASI")
print("="*70)

print("\n1ï¸âƒ£  KLASÄ°K YÃ–NTEM (5 satÄ±r):")
print("""
cift_sayilar = []
for sayi in my_numbers:
    if sayi % 2 == 0:
        cift_sayilar.append(sayi)
""")

print("2ï¸âƒ£  LIST COMPREHENSION (1 satÄ±r):")
print("""
cift_sayilar = [sayi for sayi in my_numbers if sayi % 2 == 0]
""")

print("âœ… Her iki yÃ¶ntem de aynÄ± sonucu verir!")
print("âš¡ List comprehension daha kÄ±sa ve hÄ±zlÄ±dÄ±r!")

# ============================================================================
print("\n" + "="*70)
print("ğŸ“š MODÃœL (%) OPERATÃ–RÃœ - DETAYLI AÃ‡IKLAMA")
print("="*70)

print("\nğŸ” ModÃ¼l operatÃ¶rÃ¼ Ã¶rnekleri:")
print("-" * 70)

test_sayilar = [10, 15, 24, 7, 100, 13]

for sayi in test_sayilar:
    kalan = sayi % 2
    durum = "Ã‡Ä°FT âœ“" if kalan == 0 else "TEK  âœ—"
    print(f"   {sayi:3} % 2 = {kalan}  â†’  {durum}")

# ============================================================================
print("\n" + "="*70)
print("ğŸ¯ SORU 6 Ã–ZET ve EN Ä°YÄ° PRATÄ°KLER")
print("="*70)

print("""
ğŸ“Œ Ã‡ift SayÄ± KontrolÃ¼:
   â€¢ FormÃ¼l: sayi % 2 == 0
   â€¢ Ã‡ift sayÄ±: 2, 4, 6, 8, 10, 12, ...
   â€¢ Tek sayÄ±: 1, 3, 5, 7, 9, 11, ...

ğŸ’¡ ÃœÃ§ Ã‡Ã¶zÃ¼m YÃ¶ntemi:
   1. For + if: Temel, anlaÅŸÄ±lÄ±r, eÄŸitim amaÃ§lÄ±
   2. Liste + append: SonuÃ§larÄ± saklamak iÃ§in
   3. List comprehension: Profesyonel, kÄ±sa, hÄ±zlÄ±

âš¡ Best Practice:
   â€¢ Basit iÅŸlemler â†’ List comprehension kullan
   â€¢ KarmaÅŸÄ±k iÅŸlemler â†’ Klasik for dÃ¶ngÃ¼sÃ¼
   â€¢ Okunabilirlik her zaman Ã¶nceliklidir!

ğŸ¯ Bu Ã–rnekte:
   â€¢ Liste: [1,2,3,4,5,6,19,20,32,21,20,1111,23,24]
   â€¢ Ã‡ift sayÄ±lar: 2, 4, 6, 20, 32, 20, 24
   â€¢ Toplam: 7 adet Ã§ift sayÄ±
""")

print("="*70)
print("âœ… SORU 6 TAMAMLANDI!")
print("="*70)
```

```python
#TÃ¼m dairelerin Ã§evresini iÃ§eren baÅŸka yeni bir liste oluÅŸturunuz. (Ä°pucu: 2 * pi * r)  Pi 3.14 alÄ±nabilir.
```

```python
r_list = [3,2,5,8,4,6,9,12]
```

```python
# ============================================================================
# CEVAP 7: DAÄ°RE Ã‡EVRESÄ° HESAPLAMA - LÄ°STE TRANSFORMASYONU
# ============================================================================

# ğŸ“Œ YARI Ã‡AP LÄ°STESÄ°:
r_list = [3, 2, 5, 8, 4, 6, 9, 12]

print("="*70)
print("â­• DAÄ°RE Ã‡EVRESÄ° HESAPLAMA - MATEMATÄ°KSEL LÄ°STE DÃ–NÃœÅÃœMÃœ")
print("="*70)

# ğŸ“š DAÄ°RE Ã‡EVRESÄ° FORMÃœLÃœ
print("\nğŸ“š Daire Ã‡evresi FormÃ¼lÃ¼:")
print("-" * 70)
print("Ã‡evre = 2 Ã— Ï€ Ã— r")
print("   Ï€ (pi) â‰ˆ 3.14")
print("   r = YarÄ± Ã§ap")
print("-" * 70)

# ğŸ” YARI Ã‡AP LÄ°STEMÄ°ZÄ° Ä°NCELEYELÄ°M
print(f"\nğŸ” YarÄ± Ã§ap listesi: {r_list}")
print(f"ğŸ“Š Toplam {len(r_list)} adet daire var")

# Pi deÄŸerini tanÄ±mla
pi = 3.14

print(f"\nğŸ’¡ Pi deÄŸeri: {pi}")

# ğŸ“Œ PROBLEM: Her dairenin Ã§evresini hesapla ve yeni liste oluÅŸtur
print("\n" + "="*70)
print("â“ GÃ–REV: Her dairenin Ã§evresini hesapla â†’ Yeni liste oluÅŸtur")
print("="*70)

# ğŸ”§ YÃ–NTEM 1: For DÃ¶ngÃ¼sÃ¼ ile Hesaplama (TEMEL YÃ–NTEM)
print("\nğŸ”§ YÃ–NTEM 1: For DÃ¶ngÃ¼sÃ¼ + append")
print("-" * 70)

# BoÅŸ liste oluÅŸtur
cevre_listesi = []

print("\nğŸ” Her yarÄ± Ã§ap iÃ§in Ã§evre hesaplÄ±yoruz:\n")

for r in r_list:
    cevre = 2 * pi * r
    cevre_listesi.append(cevre)
    print(f"   r = {r:2}  â†’  Ã‡evre = 2 Ã— {pi} Ã— {r:2} = {cevre:6.2f}")

print("\n" + "-" * 70)
print(f"ğŸ“‹ Ã‡evre Listesi: {cevre_listesi}")
print("-" * 70)
```

```python
# ============================================================================
# YÃ–NTEM 2: List Comprehension ile Hesaplama (PYTHONIC)
# ============================================================================

r_list = [3, 2, 5, 8, 4, 6, 9, 12]
pi = 3.14

print("\n" + "="*70)
print("ğŸ”§ YÃ–NTEM 2: List Comprehension (TEK SATIRDA!)")
print("="*70)

# List comprehension ile Ã§evre hesaplama
cevre_listesi_pythonic = [2 * pi * r for r in r_list]

print("\nğŸ“ Kod:")
print("   cevre_listesi = [2 * pi * r for r in r_list]")
print(f"\nğŸ“‹ SonuÃ§: {cevre_listesi_pythonic}")

# SonuÃ§larÄ± formatlÄ± gÃ¶ster
print("\nğŸ“Š DetaylÄ± SonuÃ§lar:")
print("-" * 70)
print("   YarÄ± Ã‡ap  â†’  Ã‡evre")
print("-" * 70)

for r, cevre in zip(r_list, cevre_listesi_pythonic):
    print(f"      {r:2}      â†’  {cevre:6.2f}")

print("-" * 70)
```

```python
# ============================================================================
# YÃ–NTEM 3: map() Fonksiyonu ile (Ä°LERÄ° SEVÄ°YE) + GÃ¶rselleÅŸtirme
# ============================================================================

r_list = [3, 2, 5, 8, 4, 6, 9, 12]
pi = 3.14

print("\n" + "="*70)
print("ğŸ”§ YÃ–NTEM 3: map() Fonksiyonu (FONKSÄ°YONEL PROGRAMLAMA)")
print("="*70)

# Lambda fonksiyonu ile map kullanÄ±mÄ±
cevre_listesi_map = list(map(lambda r: 2 * pi * r, r_list))

print("\nğŸ“ Kod:")
print("   cevre_listesi = list(map(lambda r: 2 * pi * r, r_list))")
print(f"\nğŸ“‹ SonuÃ§: {cevre_listesi_map}")

# ============================================================================
print("\n" + "="*70)
print("ğŸ“Š GÃ–RSELLEÅTÄ°RME - ASCII CHART")
print("="*70)

print("\nğŸ“ˆ YarÄ± Ã‡ap vs Ã‡evre GrafiÄŸi:")
print("-" * 70)

for r, cevre in zip(r_list, cevre_listesi_map):
    bar_length = int(cevre / 5)  # Her 5 birim iÃ§in bir #
    bar = "â–ˆ" * bar_length
    print(f"r={r:2}  {bar} {cevre:6.2f}")

print("-" * 70)

# ============================================================================
print("\n" + "="*70)
print("ğŸ¯ SORU 7 Ã–ZET ve KARÅILAÅTIRMA")
print("="*70)

print("""
ğŸ“Œ Daire Ã‡evresi FormÃ¼lÃ¼:
   â€¢ Ã‡evre = 2 Ã— Ï€ Ã— r
   â€¢ Ï€ (pi) â‰ˆ 3.14 veya 3.14159...
   â€¢ r = YarÄ± Ã§ap (radius)

ğŸ’¡ ÃœÃ§ Ã‡Ã¶zÃ¼m YÃ¶ntemi:
   1. For + append: 
      cevre_listesi = []
      for r in r_list:
          cevre_listesi.append(2 * pi * r)
   
   2. List comprehension:
      cevre_listesi = [2 * pi * r for r in r_list]
   
   3. map() fonksiyonu:
      cevre_listesi = list(map(lambda r: 2 * pi * r, r_list))

âš¡ Best Practice:
   â€¢ List comprehension: En okunabilir ve Pythonic
   â€¢ map(): Fonksiyonel programlama tarzÄ±
   â€¢ For dÃ¶ngÃ¼sÃ¼: KarmaÅŸÄ±k iÅŸlemler iÃ§in

ğŸ¯ Bu Ã–rnekte:
   â€¢ YarÄ± Ã§aplar: [3, 2, 5, 8, 4, 6, 9, 12]
   â€¢ Ã‡evreler: [18.84, 12.56, 31.40, 50.24, 25.12, 37.68, 56.52, 75.36]
""")

print("="*70)
print("âœ… SORU 7 TAMAMLANDI!")
print("="*70)
```

```python
# Sadece yaÅŸlarÄ±n olduÄŸu yeni ve ayrÄ± bir liste oluÅŸturunuz.
```

```python
age_name_list = [("Ahmet",30),("Ayse",24),("Mehmet",40),("Fatma",29)]
```

```python
# ============================================================================
# CEVAP 8: TUPLE UNPACKING - YAÅ LÄ°STESÄ° OLUÅTURMA
# ============================================================================

# ğŸ“Œ Ä°SÄ°M-YAÅ Ã‡Ä°FTLERÄ° LÄ°STESÄ°:
age_name_list = [("Ahmet",30), ("Ayse",24), ("Mehmet",40), ("Fatma",29)]

print("="*70)
print("ğŸ“¦ TUPLE UNPACKING - YAÅ LÄ°STESÄ° OLUÅTURMA")
print("="*70)

# ğŸ“š TUPLE NEDÄ°R?
print("\nğŸ“š Tuple Nedir?")
print("-" * 70)
print("Tuple: Parantez () iÃ§inde, deÄŸiÅŸtirilemez (immutable) veri yapÄ±sÄ±")
print("Ã–rnek: (\"Ahmet\", 30) â†’ isim ve yaÅŸ Ã§ifti")
print("Unpacking: Tuple'daki deÄŸerleri ayrÄ± deÄŸiÅŸkenlere atama")
print("-" * 70)

# ğŸ” LÄ°STEMÄ°ZÄ° Ä°NCELEYELÄ°M
print(f"\nğŸ” Ä°sim-YaÅŸ listesi:")
for eleman in age_name_list:
    print(f"   {eleman}  (Tip: {type(eleman).__name__})")

print(f"\nğŸ“Š Toplam {len(age_name_list)} kiÅŸi var")

# ğŸ“Œ PROBLEM: Sadece yaÅŸlarÄ± iÃ§eren yeni liste oluÅŸtur
print("\n" + "="*70)
print("â“ GÃ–REV: Sadece yaÅŸlarÄ±n olduÄŸu yeni liste oluÅŸtur")
print("="*70)

# ğŸ”§ YÃ–NTEM 1: For DÃ¶ngÃ¼sÃ¼ + Ä°ndeksleme
print("\nğŸ”§ YÃ–NTEM 1: For DÃ¶ngÃ¼sÃ¼ + Ä°ndeksleme [1]")
print("-" * 70)

yas_listesi = []

print("\nğŸ” Her tuple'dan yaÅŸÄ± alÄ±yoruz:\n")

for tuple_eleman in age_name_list:
    isim = tuple_eleman[0]  # Ä°lk eleman (indeks 0)
    yas = tuple_eleman[1]   # Ä°kinci eleman (indeks 1)
    yas_listesi.append(yas)
    print(f"   {tuple_eleman}  â†’  YaÅŸ: {yas}")

print("\n" + "-" * 70)
print(f"ğŸ“‹ YaÅŸ Listesi: {yas_listesi}")
print("-" * 70)
```

```python
# ============================================================================
# YÃ–NTEM 2: Tuple Unpacking ile (EN Ä°YÄ° YÃ–NTEM!)
# ============================================================================

age_name_list = [("Ahmet",30), ("Ayse",24), ("Mehmet",40), ("Fatma",29)]

print("\n" + "="*70)
print("ğŸ”§ YÃ–NTEM 2: Tuple Unpacking (PYTHONIC!)")
print("="*70)

yas_listesi_unpacking = []

print("\nğŸ” Tuple unpacking ile yaÅŸlarÄ± Ã§Ä±karÄ±yoruz:\n")

for isim, yas in age_name_list:  # âœ¨ Tuple unpacking!
    yas_listesi_unpacking.append(yas)
    print(f"   Ä°sim: {isim:7}  â†’  YaÅŸ: {yas}")

print("\n" + "-" * 70)
print(f"ğŸ“‹ YaÅŸ Listesi: {yas_listesi_unpacking}")
print("-" * 70)

print("\nğŸ’¡ Unpacking AÃ§Ä±klamasÄ±:")
print("   for isim, yas in age_name_list:")
print("       â†‘      â†‘")
print("   1. eleman  2. eleman")
print("\n   Python otomatik olarak tuple'Ä± aÃ§Ä±p deÄŸiÅŸkenlere atar!")
```

```python
# ============================================================================
# YÃ–NTEM 3: List Comprehension + Unpacking (EN KISA!)
# ============================================================================

age_name_list = [("Ahmet",30), ("Ayse",24), ("Mehmet",40), ("Fatma",29)]

print("\n" + "="*70)
print("ğŸ”§ YÃ–NTEM 3: List Comprehension + Unpacking")
print("="*70)

# Tek satÄ±rda Ã§Ã¶zÃ¼m!
yas_listesi_pythonic = [yas for isim, yas in age_name_list]

print("\nğŸ“ Kod (TEK SATIR!):")
print("   yas_listesi = [yas for isim, yas in age_name_list]")
print(f"\nğŸ“‹ SonuÃ§: {yas_listesi_pythonic}")

# Alternatif: Sadece indeks kullanarak
yas_listesi_indeks = [tuple_eleman[1] for tuple_eleman in age_name_list]
print(f"\nğŸ“ Alternatif (indeks ile):")
print("   yas_listesi = [tuple_eleman[1] for tuple_eleman in age_name_list]")
print(f"ğŸ“‹ SonuÃ§: {yas_listesi_indeks}")

# ============================================================================
print("\n" + "="*70)
print("ğŸ“Š Ä°STATÄ°STÄ°KLER")
print("="*70)

print(f"\nğŸ“ˆ YaÅŸ Ä°statistikleri:")
print("-" * 70)
print(f"   En kÃ¼Ã§Ã¼k yaÅŸ: {min(yas_listesi_pythonic)}")
print(f"   En bÃ¼yÃ¼k yaÅŸ: {max(yas_listesi_pythonic)}")
print(f"   Toplam yaÅŸ: {sum(yas_listesi_pythonic)}")
print(f"   Ortalama yaÅŸ: {sum(yas_listesi_pythonic) / len(yas_listesi_pythonic):.2f}")
print("-" * 70)

# ============================================================================
print("\n" + "="*70)
print("ğŸ¯ SORU 8 Ã–ZET ve TUPLE UNPACKING")
print("="*70)

print("""
ğŸ“Œ Tuple (Demet) YapÄ±sÄ±:
   â€¢ SÃ¶zdizimi: (eleman1, eleman2, ...)
   â€¢ DeÄŸiÅŸtirilemez (immutable)
   â€¢ Birden fazla deÄŸeri gruplamak iÃ§in kullanÄ±lÄ±r
   â€¢ Genelde iliÅŸkili verileri saklar

ğŸ’¡ Tuple Unpacking:
   â€¢ Tuple'daki deÄŸerleri ayrÄ± deÄŸiÅŸkenlere atama
   â€¢ for isim, yas in liste:  â† Bu unpacking!
   â€¢ Ã‡ok temiz ve okunabilir kod

ğŸ¯ ÃœÃ§ YÃ¶ntem KarÅŸÄ±laÅŸtÄ±rmasÄ±:
   1. Ä°ndeksleme: tuple_eleman[1]
      âœ“ Basit, anlaÅŸÄ±lÄ±r
      âœ— Ä°ndeks numarasÄ± anlamlÄ± deÄŸil
   
   2. Unpacking: for isim, yas in liste:
      âœ“ Ã‡ok okunabilir
      âœ“ DeÄŸiÅŸken isimleri anlamlÄ±
      âœ“ Ã–nerilen yÃ¶ntem!
   
   3. List Comprehension + Unpacking:
      âœ“ En kÄ±sa (1 satÄ±r)
      âœ“ Pythonic
      âœ“ Profesyonel

ğŸ¯ Bu Ã–rnekte:
   â€¢ Liste: [("Ahmet",30), ("Ayse",24), ("Mehmet",40), ("Fatma",29)]
   â€¢ YaÅŸlar: [30, 24, 40, 29]
   â€¢ Ortalama yaÅŸ: 30.75
""")

print("="*70)
print("âœ… SORU 8 TAMAMLANDI!")
print("="*70)
```

```python
# ============================================================================
# CEVAP 10: LIST COMPREHENSION - BOOLEAN LÄ°STESÄ° OLUÅTURMA
# ============================================================================

# ğŸ“Œ SAYILAR LÄ°STESÄ°:
number_list = [5, 7, 18, 21, 20, 10, 405, 24]

print("="*70)
print("ğŸ¯ LIST COMPREHENSION - BOOLEAN LÄ°STESÄ° (FÄ°NAL SORUSU!)")
print("="*70)

# ğŸ“š SORU ANALÄ°ZÄ°
print("\nğŸ“š Verilen Ä°fade:")
print("-" * 70)
print("   [num % 2 == 0 for num in number_list]")
print("-" * 70)

# ğŸ” Ä°FADEYÄ° PARÃ‡ALAYALIM
print("\nğŸ” Ä°fadeyi ParÃ§alara AyÄ±ralÄ±m:")
print("-" * 70)
print("   [num % 2 == 0   for num   in number_list]")
print("    â†‘               â†‘         â†‘")
print("    1. Ä°fade        2. DÃ¶ngÃ¼  3. Kaynak")
print()
print("1ï¸âƒ£  num % 2 == 0  â†’ Her sayÄ±nÄ±n Ã§ift olup olmadÄ±ÄŸÄ±nÄ± kontrol eder")
print("2ï¸âƒ£  for num       â†’ Her bir sayÄ± iÃ§in dÃ¶ngÃ¼")
print("3ï¸âƒ£  in number_list â†’ Kaynak liste")
print("-" * 70)

# ğŸ“Š LÄ°STEMÄ°ZÄ° Ä°NCELEYELÄ°M
print(f"\nğŸ“Š SayÄ±lar listesi: {number_list}")
print(f"ğŸ“Š Eleman sayÄ±sÄ±: {len(number_list)}")

# ğŸ¯ Ä°FADEYI ADIM ADIM Ã‡ALIÅTIRALIM
print("\n" + "="*70)
print("ğŸ”„ Ä°FADENÄ°N ADIM ADIM Ã‡ALIÅMASI:")
print("="*70)

print("\nğŸ” Her sayÄ± iÃ§in Ã§ift olup olmadÄ±ÄŸÄ±nÄ± kontrol ediyoruz:\n")

for num in number_list:
    sonuc = num % 2 == 0
    durum = "Ã‡Ä°FT âœ“" if sonuc else "TEK  âœ—"
    print(f"   {num:3} % 2 == 0  â†’  {num:3} % 2 = {num % 2}  â†’  {sonuc:5}  ({durum})")

# GERÃ‡EK SONUÃ‡
print("\n" + "="*70)
print("ğŸ¬ KODUN GERÃ‡EK Ã‡IKTISI:")
print("="*70)

sonuc_listesi = [num % 2 == 0 for num in number_list]

print(f"\nğŸ“ Kod: [num % 2 == 0 for num in number_list]")
print(f"\nâœ… SONUÃ‡: {sonuc_listesi}")
print("\n" + "="*70)
```

```python
# ============================================================================
# DETAYLI AÃ‡IKLAMA - List Comprehension NasÄ±l Ã‡alÄ±ÅŸÄ±r?
# ============================================================================

number_list = [5, 7, 18, 21, 20, 10, 405, 24]

print("\n" + "="*70)
print("ğŸ“– LIST COMPREHENSION DETAYLI AÃ‡IKLAMA")
print("="*70)

print("\nğŸ’¡ List Comprehension Anatomisi:")
print("-" * 70)
print("""
Genel YapÄ±: [ifade for eleman in kaynak_liste if koÅŸul]
            â†“       â†“   â†“       â†“  â†“
            1       2   3       4  5

1. ifade         : Her eleman iÃ§in uygulanacak iÅŸlem/kontrol
2. for           : DÃ¶ngÃ¼ baÅŸlangÄ±cÄ±
3. eleman        : DÃ¶ngÃ¼ deÄŸiÅŸkeni
4. in            : Ãœyelik operatÃ¶rÃ¼
5. kaynak_liste  : ElemanlarÄ±n alÄ±nacaÄŸÄ± liste
""")

print("-" * 70)

# Bu Ã¶rnekteki list comprehension
print("\nğŸ¯ Bu Ã–rnekteki List Comprehension:")
print("-" * 70)
print("   [num % 2 == 0 for num in number_list]")
print()
print("   â€¢ num % 2 == 0    : Her sayÄ±nÄ±n Ã§ift olup olmadÄ±ÄŸÄ±nÄ± kontrol et")
print("   â€¢ for num         : Her sayÄ± iÃ§in")
print("   â€¢ in number_list  : number_list'ten al")
print()
print("   SonuÃ§: Boolean deÄŸerlerden (True/False) oluÅŸan bir liste!")
print("-" * 70)

# KLASÄ°K YÃ–NTEMLE KARÅILAÅTIRMA
print("\n" + "="*70)
print("âš–ï¸  KLASÄ°K FOR DÃ–NGÃœSÃœ vs LIST COMPREHENSION")
print("="*70)

print("\n1ï¸âƒ£  KLASÄ°K YÃ–NTEM (5 satÄ±r):")
print("""
sonuc_listesi = []
for num in number_list:
    sonuc = num % 2 == 0
    sonuc_listesi.append(sonuc)
print(sonuc_listesi)
""")

# Klasik yÃ¶ntemi Ã§alÄ±ÅŸtÄ±r
sonuc_klasik = []
for num in number_list:
    sonuc = num % 2 == 0
    sonuc_klasik.append(sonuc)

print(f"SonuÃ§: {sonuc_klasik}")

print("\n2ï¸âƒ£  LIST COMPREHENSION (1 satÄ±r):")
print("""
sonuc_listesi = [num % 2 == 0 for num in number_list]
print(sonuc_listesi)
""")

# List comprehension ile
sonuc_pythonic = [num % 2 == 0 for num in number_list]

print(f"SonuÃ§: {sonuc_pythonic}")

print("\nâœ… Her iki yÃ¶ntem de aynÄ± sonucu verir!")
print("âš¡ List comprehension Ã§ok daha kÄ±sa ve okunabilir!")
```

```python
# ============================================================================
# BONUS - List Comprehension Ä°leri Seviye Ã–rnekler ve Final Ã–zet
# ============================================================================

number_list = [5, 7, 18, 21, 20, 10, 405, 24]

print("\n" + "="*70)
print("ğŸš€ BONUS: LIST COMPREHENSION Ä°LERÄ° SEVÄ°YE Ã–RNEKLER")
print("="*70)

# Ã–rnek 1: Sadece Ã§ift sayÄ±larÄ± al
print("\nğŸ“Œ Ã–RNEK 1: Sadece Ã§ift sayÄ±larÄ± al")
print("-" * 70)
cift_sayilar = [num for num in number_list if num % 2 == 0]
print(f"Kod: [num for num in number_list if num % 2 == 0]")
print(f"SonuÃ§: {cift_sayilar}")

# Ã–rnek 2: Her sayÄ±nÄ±n karesini al
print("\nğŸ“Œ Ã–RNEK 2: Her sayÄ±nÄ±n karesini al")
print("-" * 70)
kareler = [num ** 2 for num in number_list]
print(f"Kod: [num ** 2 for num in number_list]")
print(f"SonuÃ§: {kareler}")

# Ã–rnek 3: Ã‡ift sayÄ±larÄ±n iki katÄ±nÄ± al
print("\nğŸ“Œ Ã–RNEK 3: Sadece Ã§ift sayÄ±larÄ±n iki katÄ±nÄ± al")
print("-" * 70)
cift_iki_kat = [num * 2 for num in number_list if num % 2 == 0]
print(f"Kod: [num * 2 for num in number_list if num % 2 == 0]")
print(f"SonuÃ§: {cift_iki_kat}")

# Ã–rnek 4: Ternary operatÃ¶r ile
print("\nğŸ“Œ Ã–RNEK 4: Ã‡ift ise 'Ã‡Ä°FT', tek ise 'TEK' yaz")
print("-" * 70)
etiketler = ['Ã‡Ä°FT' if num % 2 == 0 else 'TEK' for num in number_list]
print(f"Kod: ['Ã‡Ä°FT' if num % 2 == 0 else 'TEK' for num in number_list]")
print(f"SonuÃ§: {etiketler}")

# ============================================================================
print("\n" + "="*70)
print("ğŸ“Š SONUÃ‡LARI GÃ–RSELLEÅTÄ°RELÄ°M")
print("="*70)

print("\nğŸ“ˆ SayÄ± â†’ Boolean EÅŸleÅŸmesi:")
print("-" * 70)

sonuc_listesi = [num % 2 == 0 for num in number_list]

for num, is_even in zip(number_list, sonuc_listesi):
    durum = "Ã‡Ä°FT âœ“" if is_even else "TEK  âœ—"
    bar = "â–ˆ" * (num // 10) if num < 100 else "â–ˆ" * 10
    print(f"   {num:3}  {bar:10}  â†’  {is_even:5}  ({durum})")

print("-" * 70)

# Ä°statistikler
cift_sayisi = sonuc_listesi.count(True)
tek_sayisi = sonuc_listesi.count(False)

print(f"\nğŸ“Š Ã–zet Ä°statistikler:")
print(f"   â€¢ Ã‡ift sayÄ± adedi: {cift_sayisi}")
print(f"   â€¢ Tek sayÄ± adedi: {tek_sayisi}")
print(f"   â€¢ Toplam: {len(sonuc_listesi)}")

# ============================================================================
print("\n" + "="*70)
print("ğŸ¯ SORU 10 ve LIST COMPREHENSION Ã–ZET")
print("="*70)

print("""
ğŸ“Œ Soru 10 CevabÄ±:
   â€¢ Ä°fade: [num % 2 == 0 for num in number_list]
   â€¢ SonuÃ§: [False, False, True, False, True, True, False, True]
   â€¢ AÃ§Ä±klama: Her sayÄ± iÃ§in Ã§ift mi diye kontrol ediliyor
   â€¢ Ã‡ift sayÄ±lar iÃ§in True, tek sayÄ±lar iÃ§in False dÃ¶ndÃ¼rÃ¼lÃ¼yor

ğŸ’¡ List Comprehension Nedir?
   â€¢ Python'un en gÃ¼Ã§lÃ¼ Ã¶zelliklerinden biri
   â€¢ Tek satÄ±rda liste oluÅŸturma yÃ¶ntemi
   â€¢ For dÃ¶ngÃ¼sÃ¼nÃ¼n kompakt hali
   â€¢ Pythonic kod yazmanÄ±n temel taÅŸÄ±

ğŸ¯ List Comprehension KullanÄ±m SenaryolarÄ±:
   1. Listeyi filtreleme: [x for x in liste if koÅŸul]
   2. Listeyi dÃ¶nÃ¼ÅŸtÃ¼rme: [x*2 for x in liste]
   3. Boolean liste oluÅŸturma: [x > 5 for x in liste]
   4. Ä°Ã§ iÃ§e listeler: [x for sublist in liste for x in sublist]

âš¡ AvantajlarÄ±:
   â€¢ Daha az kod (1 satÄ±r vs 5 satÄ±r)
   â€¢ Daha hÄ±zlÄ± Ã§alÄ±ÅŸÄ±r (C seviyesinde optimize)
   â€¢ Daha okunabilir (alÄ±ÅŸÄ±nca)
   â€¢ Pythonic (profesyonel Python kodu)

âš ï¸  Ne Zaman KullanmalÄ±?
   âœ… Basit transformasyonlar iÃ§in
   âœ… Filtreleme iÅŸlemleri iÃ§in
   âœ… Boolean listeler iÃ§in
   âŒ KarmaÅŸÄ±k mantÄ±k iÃ§in (klasik for kullan)
   âŒ Yan etkiler (side effects) iÃ§in

ğŸ“ Bu Ã–rnekte Ã–ÄŸrendiklerimiz:
   â€¢ Liste: [5, 7, 18, 21, 20, 10, 405, 24]
   â€¢ Ä°fade: num % 2 == 0
   â€¢ SonuÃ§: Boolean liste (True/False)
   â€¢ Her sayÄ±nÄ±n Ã§ift olup olmadÄ±ÄŸÄ±nÄ± kontrol ettik!
""")

print("="*70)
print("âœ… SORU 10 TAMAMLANDI!")
print("="*70)
print("\n" + "ğŸ‰"*25)
print("ğŸ† TÃœM QUIZ TAMAMLANDI! MUHTEÅEM BÄ°R Ä°Å Ã‡IKARDIN! ğŸ†")
print("ğŸ‰"*25)
print("\n" + "="*70)
print("ğŸ“š QUIZ 2 - Ã–ZET")
print("="*70)
print("""
âœ… Soru 1-2:  MantÄ±ksal OperatÃ¶rler (AND, OR)
âœ… Soru 3:    KoÅŸullu Ä°fadeler (if-elif-else)
âœ… Soru 4:    Dictionary - DeÄŸer KontrolÃ¼
âœ… Soru 5:    Dictionary - Anahtar KontrolÃ¼
âœ… Soru 6:    Liste Filtreleme (Ã‡ift SayÄ±lar)
âœ… Soru 7:    Daire Ã‡evresi Hesaplama
âœ… Soru 8:    Tuple Unpacking
âœ… Soru 9:    Random ModÃ¼lÃ¼
âœ… Soru 10:   List Comprehension (Boolean)

ğŸŒŸ Python'da Ã¶ÄŸrendiÄŸin tÃ¼m temel konularÄ± uyguladÄ±n!
ğŸš€ ArtÄ±k bir Python ninja'sÄ±sÄ±n! ğŸ¥·
ğŸ’ª Devam et, daha neler Ã¶ÄŸreneceksin!
""")
print("="*70)
```
